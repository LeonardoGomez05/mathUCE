# **Funciones Inyectivas**

Una funci칩n $f: A \to B$ es **inyectiva** si cada elemento del dominio se asigna a un 칰nico elemento del codominio, evitando duplicados en el resultado.

![Texto alternativo](https://www.lifeder.com/wp-content/uploads/2019/04/funcion-inyectiva.jpg)

## **Caracter칤sticas y Propiedades** 游늳

- **Definici칩n Formal:** Una funci칩n es inyectiva si $f(x_1) = f(x_2)$ implica que $x_1 = x_2$.
  $$ f(x_1) = f(x_2) \Rightarrow x_1 = x_2 $$

- **M칠todo Gr치fico:** En una gr치fica, una funci칩n es inyectiva si ninguna l칤nea vertical atraviesa la gr치fica m치s de una vez. 游늵

- **Propiedades Adicionales:**
  - Toda funci칩n inyectiva tiene una funci칩n inversa 칰nica.
  - La composici칩n de dos funciones inyectivas es tambi칠n inyectiva. 游댌

## **Ejemplos y Contrastes** 游륲n
- **Ejemplo Inyectivo:**
  $$ f(x) = 3x + 1 $$
  Cada valor de $(x)$ produce un 칰nico valor de $(y)$.

- **Ejemplo No Inyectivo:**
  $$ g(x) = x^2 $$
  Varios valores de $(x)$ pueden dar el mismo valor de $(y)$, lo que no cumple con la condici칩n de inyectividad.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definici칩n de la funci칩n inyectiva
def f(x):
    return 2 * x + 1

# Generaci칩n de datos para la gr치fica
x = np.linspace(-5, 5, 100)  # Genera 100 puntos entre -5 y 5
y = f(x)                     # Aplica la funci칩n a todos los puntos x

# Configuraci칩n de la gr치fica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 2x + 1', color='b', linewidth=2)  # Grafica f(x)
plt.axhline(0, color='black',linewidth=0.5)  # Linea horizontal eje X
plt.axvline(0, color='black',linewidth=0.5)  # Linea vertical eje Y
plt.title('Gr치fica de una Funci칩n Inyectiva')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Marca un punto espec칤fico para demostrar la inyectividad
plt.scatter(2, f(2), color='r', label='Punto Inyectivo (2, 5)', zorder=5)
plt.annotate('(2, 5)', xy=(2, f(2)), xytext=(3, f(2) + 2),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Mostrar la gr치fica con ejes visibles
plt.axhline(0, color='black',linewidth=0.5)  # Linea horizontal eje X
plt.axvline(0, color='black',linewidth=0.5)  # Linea vertical eje Y
plt.show()
```

# **Funciones Sobreyectivas** 游꿢

Una funci칩n $f: A \to B$ es **sobreyectiva** si cada elemento del conjunto $B$ es alcanzado al menos una vez por alg칰n elemento del conjunto $A$. En otras palabras, cada elemento de $B$ tiene al menos una preimagen en $A$.

![Texto alternativo](https://www.lifeder.com/wp-content/uploads/2019/05/funcion-sobreyectiva.jpg)

## **Definici칩n Formal** 游닇

Una funci칩n $f: A\to B$ es sobreyectiva si, para todo $y \in B$, existe al menos un $x \in A$ tal que $f(x) = y$.

## **Ejemplo** 游

Consideremos la funci칩n \$ f: \mathbb{R} \to \mathbb{R} \$ definida por:

$$
f(x) = x^2
$$

Vamos a demostrar que es sobreyectiva:

Para cualquier $y \in \mathbb{R}$, podemos elegir $x = \sqrt{y}$ si $y \geq 0$, o $x = -\sqrt{y}$ si $y < 0$. En ambos casos, $f(x) = y$, lo que demuestra que $f$ es sobreyectiva.

## **Propiedades de las Funciones Sobreyectivas** 游륲n
1. **Composici칩n de Funciones**: Si $f: A \to B$ y $g: B \to C$ son sobreyectivas, entonces la composici칩n $g \circ f: A \to C$ es tambi칠n sobreyectiva.
2. **Inversa a la Izquierda**: Una funci칩n $f: A \to B$ es sobreyectiva si y solo si tiene una inversa a la izquierda, es decir, existe una funci칩n $g: B \to A$ tal que $g(f(x)) = x$ para todo $x \in A$.

## **Ejemplos Adicionales de Funciones Sobreyectivas** 游륲n
### **Ejemplo 1: Funci칩n Lineal**
Considera la funci칩n $f: \mathbb{R} \to \mathbb{R}$ definida por $f(x) = 2x + 1$. Esta funci칩n es sobreyectiva porque para cualquier $y \in \mathbb{R}$, podemos despejar $x$ como $x = \frac{y - 1}{2}$, asegurando que cada $y$ tenga al menos una preimagen en $\mathbb{R}$.

#### **Ejemplo 2: Funci칩n Exponencial**
La funci칩n $f: \mathbb{R} \to (0, \infty)$ dada por $f(x) = e^x$ es sobreyectiva. Cada n칰mero positivo $y \in (0, \infty)$ tiene al menos una preimagen $x \in \mathbb{R}$ tal que $ e^x = y$.

## **Funci칩n Biyectiva**游游댌

### **Definici칩n** 游닇

Una funci칩n $ f: A \to B $ es **biyectiva**, si cumple con las siguientes dos propiedades:

1. **Inyectividad**: $ f $ es inyectiva si $ f(a_1) = f(a_2) \implies a_1 = a_2 $ para todo $ a_1, a_2 \in A $. Esto significa que no hay dos elementos distintos en el dominio $ A $ que se mapeen al mismo elemento en el codominio $ B $.
   
2. **Sobreyectividad**: $ f $ es sobreyectiva si para todo $ b \in B $, existe al menos un $ a \in A $ tal que $ f(a) = b $. Esto significa que cada elemento del codominio $ B $ tiene al menos un preimagen en el dominio $ A $.

Cuando una funci칩n es tanto inyectiva como sobreyectiva, se dice que es **biyectiva**. En otras palabras, cada elemento del codominio est치 relacionado con exactamente un elemento del dominio, y viceversa.

### **Propiedades de las Funciones Biyectivas** 游륲n
1. **Existencia de Inversa**: Si una funci칩n $ f $ es biyectiva, entonces existe una funci칩n inversa $ f^{-1} $ tal que $ f(f^{-1}(b)) = b $ para todo $ b \in B $ y $ f^{-1}(f(a)) = a $ para todo $ a \in A $.

2. **Cardinalidad**: Si $ f: A \to B $ es una biyectiva, entonces el conjunto $ A $ y el conjunto $ B $ tienen la misma cardinalidad, es decir, el mismo n칰mero de elementos.

### **Ejemplos de Funciones Biyectivas** 游닄

1. **Funci칩n Lineal**:
   Sea $ f: \mathbb{R} \to \mathbb{R} $ definida por $ f(x) = 2x + 3 $. Demostraremos que es biyectiva.

   **Inyectividad**:
   Supongamos $ f(a_1) = f(a_2) $. Entonces,

   $2a_1 + 3 = 2a_2 + 3 \implies 2a_1 = 2a_2 \implies a_1 = a_2$

   **Sobreyectividad**:
   Para cualquier $ y \in \mathbb{R} $, necesitamos encontrar $ x \in \mathbb{R} $ tal que $ f(x) = y $.
   $y = 2x + 3 \implies x = \frac{y - 3}{2}$
   Por lo tanto, para cualquier $ y $, $ x = \frac{y - 3}{2} $ est치 en $\mathbb{R}$, lo que prueba la sobreyectividad.

2. **Funci칩n Exponencial**:
   Sea $ g: \mathbb{R} \to \mathbb{R}^+ $ definida por $ g(x) = e^x $.

   **Inyectividad**:
   Supongamos $ g(a_1) = g(a_2) $. Entonces,
   
   $e^{a_1} = e^{a_2} \implies a_1 = a_2$

   **Sobreyectividad**:
   Para cualquier $ y \in \mathbb{R}^+ $, necesitamos encontrar $ x \in \mathbb{R} $ tal que $ g(x) = y $.
   
   $y = e^x \implies x = \ln(y)$

   Dado que $ \ln(y) \in \mathbb{R} $ para todo $ y \in \mathbb{R}^+ $, $g$ es sobreyectiva.

### **Aplicaciones de las Funciones Biyectivas** 游깷

1. **Criptograf칤a**: En criptograf칤a, las funciones biyectivas se utilizan para crear cifrados que mapean cada mensaje original a un 칰nico mensaje cifrado y viceversa, permitiendo tanto la codificaci칩n como la decodificaci칩n segura de mensajes.

# **Funci칩n Inversa** 游댃

## **Definici칩n y Propiedades**
Una funci칩n inversa $ f^{-1} $ de una funci칩n $ f $ existe si y solo si $ f $ es biyectiva. Esto significa que cada elemento en el conjunto de llegada (imagen) de $ f $ est치 asociado de manera 칰nica con un elemento en el conjunto de salida (dominio) de $ f $, y viceversa.

![Texto alternativo](https://th.bing.com/th/id/OIP.qprm5lwJfL4K_96wmdtITwAAAA?rs=1&pid=ImgDetMain)

- **Existencia**: Una funci칩n $ f $ tiene una funci칩n inversa si es biyectiva. Esto implica que $ f $ es tanto inyectiva (cada elemento de $ A $ se mapea a un 칰nico elemento de $ B $) como sobreyectiva (cada elemento de $ B $ tiene al menos un preimagen en $ A $).

- **Propiedad de Inversi칩n**: Si $ f $ tiene una funci칩n inversa $ f^{-1} $, entonces para cada $ y $ en el dominio de $ f $ y para cada $ x $ en el rango de $ f $, se cumple que $ f(f^{-1}(y)) = y $ y $ f^{-1}(f(x)) = x $.

## **C치lculo de la Funci칩n Inversa** 游댝

Para encontrar la funci칩n inversa $ f^{-1} $ de una funci칩n $ f $:

1. **Expresi칩n de $ y $ en t칠rminos de $ x $**: Si $ f(x) = y $, se despeja $ x $ en t칠rminos de $ y $.

2. **Interpretaci칩n de la Inversa**: La funci칩n inversa $ f^{-1}(y) $ se obtiene intercambiando las variables $ x $ e $ y $ en la ecuaci칩n original.

## **Ejemplo Detallado** 游닇

Supongamos $ f(x) = 2x + 3 $:

- **Paso 1**: Expresamos $ y $ en t칠rminos de $ x $:
  $ y = 2x + 3 $

- **Paso 2**: Despejamos $ x $ en t칠rminos de $ y $:
  $ y - 3 = 2x $
  $ x = \frac{y - 3}{2} $

- **Paso 3**: Definimos $ f^{-1}(y) $ como:
  $ f^{-1}(y) = \frac{y - 3}{2} $

## **Aplicaciones Avanzadas** 游깷

Las funciones inversas tienen numerosas aplicaciones en 치reas como criptograf칤a, ingenier칤a, econom칤a y m치s. Son fundamentales para modelar relaciones inversamente proporcionales y resolver problemas matem치ticos complejos.

#### **Importancia Te칩rica y Pr치ctica** 游

La comprensi칩n de las funciones inversas no solo es crucial para la matem치tica pura, sino que tambi칠n tiene aplicaciones pr치cticas significativas en diversas disciplinas cient칤ficas y t칠cnicas.

```python
import matplotlib.pyplot as plt
import numpy as np

# Definici칩n de la funci칩n original
def f(x):
    """
    Funci칩n original f(x) = 2x + 3
    """
    return 2*x + 3

# Definici칩n de la funci칩n inversa
def inverse_f(y):
    """
    Funci칩n inversa de f(x) = 2x + 3
    """
    return (y - 3) / 2

# Generar datos para la gr치fica
x = np.linspace(-10, 10, 400)
y = f(x)
inverse_y = inverse_f(x)

# Graficar la funci칩n y su inversa
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = 2x + 3', color='blue')
plt.plot(x, inverse_y, label='f^{-1}(x)', color='orange')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Funci칩n original y su inversa')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

# **Funci칩n Compuesta: Definici칩n y Propiedades**游댃

La funci칩n compuesta combina dos funciones para formar una nueva funci칩n $(f \circ g)(x)$, donde $f$ y $g$ son funciones. Se define como la aplicaci칩n de $f$ sobre el resultado de $g(x)$.

![Texto alternativo](https://www.funciones.xyz/wp-content/uploads/2021/03/composicion-de-funciones-300x109.png?ezimgfmt=ng:webp/ngcb1)

## **F칩rmula de la Funci칩n Compuesta** 游닇

La f칩rmula general para la funci칩n compuesta de dos funciones $f$ y $g$ se expresa como:
$$(f \circ g)(x) = f(g(x))$$

![Texto alternativo](https://th.bing.com/th/id/R.1c3570ad60516e5184f4c760f895c6be?rik=xHhIJ4MasccAJQ&pid=ImgRaw&r=0&sres=1&sresct=1)

## **Propiedades y Caracter칤sticas** 游댌

Las funciones compuestas poseen varias propiedades importantes:

- **Asociatividad**: Las funciones compuestas son asociativas, es decir, $(f \circ g) \circ h = f \circ (g \circ h)$.
- **Identidad**: Existe una funci칩n identidad $I(x) = x$ tal que $(f \circ I)(x) = f(x)$.

![Texto alternativo](https://th.bing.com/th/id/R.1ff5b905924ba52c6aaaf29b503d8713?rik=6W8UYtX42NfG0A&pid=ImgRaw&r=0)

## **Ejemplo de Funci칩n Compuesta** 游늵

Supongamos las funciones $f(x) = 2x + 1$ y $g(x) = x^2$. La funci칩n compuesta $(f \circ g)(x)$ se calcula como:
$$(f \circ g)(x) = f(g(x)) = 2x^2 + 1$$

Comprender a fondo las funciones compuestas no solo es esencial para la matem치tica avanzada, sino tambi칠n para su aplicaci칩n pr치ctica en disciplinas como la f칤sica, econom칤a e inform치tica. Su versatilidad y capacidad para modelar sistemas din치micos y relaciones complejas las convierten en una herramienta poderosa en el an치lisis y la resoluci칩n de problemas reales.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definici칩n de las funciones individuales g(x) y f(x)
def g(x):
    return x**2  # Funci칩n g(x) = x^2

def f(x):
    return np.sqrt(x)  # Funci칩n f(x) = sqrt(x)

# Definici칩n de la funci칩n compuesta f(g(x))
def composite_function(x):
    return f(g(x))  # Funci칩n compuesta f(g(x)) = f(g(x))

# Crear un rango de valores de x para graficar
x = np.linspace(0, 3, 400)

# Calcular los valores de la funci칩n compuesta para cada valor de x
y = composite_function(x)

# Graficar la funci칩n compuesta
plt.figure(figsize=(8, 6))  # Tama침o del gr치fico
plt.plot(x, y, label='$f(g(x)) = \sqrt{x^2}$', color='b')  # Graficar f(g(x)), etiqueta en LaTeX
plt.title('Funci칩n Compuesta $f(g(x))$')  # T칤tulo del gr치fico
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('f(g(x))')  # Etiqueta del eje y
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar cuadr칤cula en el gr치fico
plt.show()  # Mostrar el gr치fico
```

#### **Conclusi칩n** 游꿉
La funci칩n de utilidad es un concepto fundamental en las ciencias sociales, especialmente en econom칤a, para modelar las preferencias y el comportamiento de los individuos. Su aplicaci칩n abarca desde la teor칤a del consumidor y la elecci칩n racional hasta la evaluaci칩n de pol칤ticas p칰blicas y el an치lisis de bienestar social.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci칩n de utilidad Cobb-Douglas
def utilidad(x, y, alpha=0.5):
    return (x**alpha) * (y**(1-alpha))

# Crear valores de X e Y para graficar las curvas de indiferencia
x = np.linspace(0.1, 10, 400)
y = np.linspace(0.1, 10, 400)
X, Y = np.meshgrid(x, y)

# Calcular la utilidad para diferentes combinaciones de X e Y
U = utilidad(X, Y)

# Definir niveles de utilidad para las curvas de indiferencia
niveles_utilidad = [1, 2, 3, 4, 5]

# Crear el gr치fico
plt.figure(figsize=(10, 6))

# Graficar las curvas de indiferencia
contornos = plt.contour(X, Y, U, levels=niveles_utilidad, linewidths=2, alpha=0.75, linestyles='dashed')
plt.clabel(contornos, inline=1, fontsize=10)

# A침adir etiquetas y t칤tulo
plt.xlabel('Bien X')
plt.ylabel('Bien Y')
plt.title('Curvas de Indiferencia para una Funci칩n de Utilidad Cobb-Douglas')
plt.colorbar(label='Utilidad')

# Mostrar el gr치fico
plt.grid(True)
plt.show()
```
