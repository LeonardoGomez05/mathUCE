# **Funciones Inyectivas**

Una función $f: A \to B$ es **inyectiva** si cada elemento del dominio se asigna a un único elemento del codominio, evitando duplicados en el resultado.

![Texto alternativo](https://www.lifeder.com/wp-content/uploads/2019/04/funcion-inyectiva.jpg)

## **Características y Propiedades** 📈

- **Definición Formal:** Una función es inyectiva si $f(x_1) = f(x_2)$ implica que $x_1 = x_2$.
  $$ f(x_1) = f(x_2) \Rightarrow x_1 = x_2 $$

- **Método Gráfico:** En una gráfica, una función es inyectiva si ninguna línea vertical atraviesa la gráfica más de una vez. 📊

- **Propiedades Adicionales:**
  - Toda función inyectiva tiene una función inversa única.
  - La composición de dos funciones inyectivas es también inyectiva. 🔍

## **Ejemplos y Contrastes** 🌟

- **Ejemplo Inyectivo:**
  $$ f(x) = 3x + 1 $$
  Cada valor de $(x)$ produce un único valor de $(y)$.

- **Ejemplo No Inyectivo:**
  $$ g(x) = x^2 $$
  Varios valores de $(x)$ pueden dar el mismo valor de $(y)$, lo que no cumple con la condición de inyectividad.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función inyectiva
def f(x):
    return 2 * x + 1

# Generación de datos para la gráfica
x = np.linspace(-5, 5, 100)  # Genera 100 puntos entre -5 y 5
y = f(x)                     # Aplica la función a todos los puntos x

# Configuración de la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 2x + 1', color='b', linewidth=2)  # Grafica f(x)
plt.axhline(0, color='black',linewidth=0.5)  # Linea horizontal eje X
plt.axvline(0, color='black',linewidth=0.5)  # Linea vertical eje Y
plt.title('Gráfica de una Función Inyectiva')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Marca un punto específico para demostrar la inyectividad
plt.scatter(2, f(2), color='r', label='Punto Inyectivo (2, 5)', zorder=5)
plt.annotate('(2, 5)', xy=(2, f(2)), xytext=(3, f(2) + 2),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Mostrar la gráfica con ejes visibles
plt.axhline(0, color='black',linewidth=0.5)  # Linea horizontal eje X
plt.axvline(0, color='black',linewidth=0.5)  # Linea vertical eje Y
plt.show()
```

# **Funciones Sobreyectivas** 🎯

Una función $f: A \to B$ es **sobreyectiva** si cada elemento del conjunto $B$ es alcanzado al menos una vez por algún elemento del conjunto $A$. En otras palabras, cada elemento de $B$ tiene al menos una preimagen en $A$.

![Texto alternativo](https://www.lifeder.com/wp-content/uploads/2019/05/funcion-sobreyectiva.jpg)

## **Definición Formal** 📝

Una función $f: A\to B$ es sobreyectiva si, para todo $y \in B$, existe al menos un $x \in A$ tal que $f(x) = y$.

## **Ejemplo** 🚀

Consideremos la función \$ f: \mathbb{R} \to \mathbb{R} \$ definida por:

$$
f(x) = x^2
$$

Vamos a demostrar que es sobreyectiva:

Para cualquier $y \in \mathbb{R}$, podemos elegir $x = \sqrt{y}$ si $y \geq 0$, o $x = -\sqrt{y}$ si $y < 0$. En ambos casos, $f(x) = y$, lo que demuestra que $f$ es sobreyectiva.

## **Propiedades de las Funciones Sobreyectivas** 🌟

1. **Composición de Funciones**: Si $f: A \to B$ y $g: B \to C$ son sobreyectivas, entonces la composición $g \circ f: A \to C$ es también sobreyectiva.
2. **Inversa a la Izquierda**: Una función $f: A \to B$ es sobreyectiva si y solo si tiene una inversa a la izquierda, es decir, existe una función $g: B \to A$ tal que $g(f(x)) = x$ para todo $x \in A$.

## **Ejemplos Adicionales de Funciones Sobreyectivas** 🌟

### **Ejemplo 1: Función Lineal**
Considera la función $f: \mathbb{R} \to \mathbb{R}$ definida por $f(x) = 2x + 1$. Esta función es sobreyectiva porque para cualquier $y \in \mathbb{R}$, podemos despejar $x$ como $x = \frac{y - 1}{2}$, asegurando que cada $y$ tenga al menos una preimagen en $\mathbb{R}$.

#### **Ejemplo 2: Función Exponencial**
La función $f: \mathbb{R} \to (0, \infty)$ dada por $f(x) = e^x$ es sobreyectiva. Cada número positivo $y \in (0, \infty)$ tiene al menos una preimagen $x \in \mathbb{R}$ tal que $ e^x = y$.

## **Función Biyectiva**🌟🔍

### **Definición** 📝

Una función $ f: A \to B $ es **biyectiva**, si cumple con las siguientes dos propiedades:

1. **Inyectividad**: $ f $ es inyectiva si $ f(a_1) = f(a_2) \implies a_1 = a_2 $ para todo $ a_1, a_2 \in A $. Esto significa que no hay dos elementos distintos en el dominio $ A $ que se mapeen al mismo elemento en el codominio $ B $.
   
2. **Sobreyectividad**: $ f $ es sobreyectiva si para todo $ b \in B $, existe al menos un $ a \in A $ tal que $ f(a) = b $. Esto significa que cada elemento del codominio $ B $ tiene al menos un preimagen en el dominio $ A $.

Cuando una función es tanto inyectiva como sobreyectiva, se dice que es **biyectiva**. En otras palabras, cada elemento del codominio está relacionado con exactamente un elemento del dominio, y viceversa.

### **Propiedades de las Funciones Biyectivas** 🌟

1. **Existencia de Inversa**: Si una función $ f $ es biyectiva, entonces existe una función inversa $ f^{-1} $ tal que $ f(f^{-1}(b)) = b $ para todo $ b \in B $ y $ f^{-1}(f(a)) = a $ para todo $ a \in A $.

2. **Cardinalidad**: Si $ f: A \to B $ es una biyectiva, entonces el conjunto $ A $ y el conjunto $ B $ tienen la misma cardinalidad, es decir, el mismo número de elementos.

### **Ejemplos de Funciones Biyectivas** 📚

1. **Función Lineal**:
   Sea $ f: \mathbb{R} \to \mathbb{R} $ definida por $ f(x) = 2x + 3 $. Demostraremos que es biyectiva.

   **Inyectividad**:
   Supongamos $ f(a_1) = f(a_2) $. Entonces,

   $2a_1 + 3 = 2a_2 + 3 \implies 2a_1 = 2a_2 \implies a_1 = a_2$

   **Sobreyectividad**:
   Para cualquier $ y \in \mathbb{R} $, necesitamos encontrar $ x \in \mathbb{R} $ tal que $ f(x) = y $.
   $y = 2x + 3 \implies x = \frac{y - 3}{2}$
   Por lo tanto, para cualquier $ y $, $ x = \frac{y - 3}{2} $ está en $\mathbb{R}$, lo que prueba la sobreyectividad.

2. **Función Exponencial**:
   Sea $ g: \mathbb{R} \to \mathbb{R}^+ $ definida por $ g(x) = e^x $.

   **Inyectividad**:
   Supongamos $ g(a_1) = g(a_2) $. Entonces,
   
   $e^{a_1} = e^{a_2} \implies a_1 = a_2$

   **Sobreyectividad**:
   Para cualquier $ y \in \mathbb{R}^+ $, necesitamos encontrar $ x \in \mathbb{R} $ tal que $ g(x) = y $.
   
   $y = e^x \implies x = \ln(y)$

   Dado que $ \ln(y) \in \mathbb{R} $ para todo $ y \in \mathbb{R}^+ $, $g$ es sobreyectiva.

### **Aplicaciones de las Funciones Biyectivas** 🌐

1. **Criptografía**: En criptografía, las funciones biyectivas se utilizan para crear cifrados que mapean cada mensaje original a un único mensaje cifrado y viceversa, permitiendo tanto la codificación como la decodificación segura de mensajes.

![Texto alternativo](https://th.bing.com/th/id/R.1f1c070f80c4463c3eb0311be6c3fd6a?rik=%2fYxjjpayP0lQeg&pid=ImgRaw&r=0&sres=1&sresct=1)

2. **Bases de Datos**: En la normalización de bases de datos, las funciones biyectivas se utilizan para asegurar que cada clave primaria tiene una correspondencia única con una tupla en la base de datos, eliminando redundancias y garantizando integridad referencial.

![Texto alternativo](https://th.bing.com/th/id/OIP.7RfPFoajfCsHaWXLgO5rCgHaHa?w=512&h=512&rs=1&pid=ImgDetMain)

3. **Teoría de Grafos**: En la teoría de grafos, una correspondencia biyectiva entre los vértices de dos grafos indica que los grafos son isomorfos, lo que significa que tienen estructuras idénticas.

![Texto alternativo](https://th.bing.com/th/id/OIP.xtKQID9zDtY8Jdma1sS1bQHaFj?rs=1&pid=ImgDetMain)

4. **Economía**: En la teoría económica, las funciones biyectivas pueden representar relaciones entre variables económicas donde cada variable independiente tiene una única variable dependiente correspondiente, facilitando el análisis de sistemas económicos complejos.

![Texto alternativo](https://www.ipe.org.pe/portal/wp-content/uploads/2012/03/economia-2018.jpg)


```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función biyectiva
def f(x):
    return np.log(x)

# Generar datos para graficar
x = np.linspace(0.1, 5, 400)  # Genera 400 puntos entre 0.1 y 5 (evitando x=0 para logaritmo)
y = f(x)

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = \ln(x)$')  # Etiqueta con la función

# Añadir detalles al gráfico
plt.title('Gráfica de la función biyectiva $f(x) = \ln(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Mostrar el gráfico
plt.show()
```

# **Función Inversa** 🔄

## **Definición y Propiedades**
Una función inversa $ f^{-1} $ de una función $ f $ existe si y solo si $ f $ es biyectiva. Esto significa que cada elemento en el conjunto de llegada (imagen) de $ f $ está asociado de manera única con un elemento en el conjunto de salida (dominio) de $ f $, y viceversa.

![Texto alternativo](https://th.bing.com/th/id/OIP.qprm5lwJfL4K_96wmdtITwAAAA?rs=1&pid=ImgDetMain)

- **Existencia**: Una función $ f $ tiene una función inversa si es biyectiva. Esto implica que $ f $ es tanto inyectiva (cada elemento de $ A $ se mapea a un único elemento de $ B $) como sobreyectiva (cada elemento de $ B $ tiene al menos un preimagen en $ A $).

- **Propiedad de Inversión**: Si $ f $ tiene una función inversa $ f^{-1} $, entonces para cada $ y $ en el dominio de $ f $ y para cada $ x $ en el rango de $ f $, se cumple que $ f(f^{-1}(y)) = y $ y $ f^{-1}(f(x)) = x $.

## **Cálculo de la Función Inversa** 🔢

Para encontrar la función inversa $ f^{-1} $ de una función $ f $:

1. **Expresión de $ y $ en términos de $ x $**: Si $ f(x) = y $, se despeja $ x $ en términos de $ y $.

2. **Interpretación de la Inversa**: La función inversa $ f^{-1}(y) $ se obtiene intercambiando las variables $ x $ e $ y $ en la ecuación original.

## **Ejemplo Detallado** 📝

Supongamos $ f(x) = 2x + 3 $:

- **Paso 1**: Expresamos $ y $ en términos de $ x $:
  $ y = 2x + 3 $

- **Paso 2**: Despejamos $ x $ en términos de $ y $:
  $ y - 3 = 2x $
  $ x = \frac{y - 3}{2} $

- **Paso 3**: Definimos $ f^{-1}(y) $ como:
  $ f^{-1}(y) = \frac{y - 3}{2} $

## **Aplicaciones Avanzadas** 🌐

Las funciones inversas tienen numerosas aplicaciones en áreas como criptografía, ingeniería, economía y más. Son fundamentales para modelar relaciones inversamente proporcionales y resolver problemas matemáticos complejos.

#### **Importancia Teórica y Práctica** 🧠

La comprensión de las funciones inversas no solo es crucial para la matemática pura, sino que también tiene aplicaciones prácticas significativas en diversas disciplinas científicas y técnicas.

```python
import matplotlib.pyplot as plt
import numpy as np

# Definición de la función original
def f(x):
    """
    Función original f(x) = 2x + 3
    """
    return 2*x + 3

# Definición de la función inversa
def inverse_f(y):
    """
    Función inversa de f(x) = 2x + 3
    """
    return (y - 3) / 2

# Generar datos para la gráfica
x = np.linspace(-10, 10, 400)
y = f(x)
inverse_y = inverse_f(x)

# Graficar la función y su inversa
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = 2x + 3', color='blue')
plt.plot(x, inverse_y, label='f^{-1}(x)', color='orange')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Función original y su inversa')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

# **Función Compuesta: Definición y Propiedades**🔄

La función compuesta combina dos funciones para formar una nueva función $(f \circ g)(x)$, donde $f$ y $g$ son funciones. Se define como la aplicación de $f$ sobre el resultado de $g(x)$.

![Texto alternativo](https://www.funciones.xyz/wp-content/uploads/2021/03/composicion-de-funciones-300x109.png?ezimgfmt=ng:webp/ngcb1)

## **Fórmula de la Función Compuesta** 📝

La fórmula general para la función compuesta de dos funciones $f$ y $g$ se expresa como:
$$(f \circ g)(x) = f(g(x))$$

![Texto alternativo](https://th.bing.com/th/id/R.1c3570ad60516e5184f4c760f895c6be?rik=xHhIJ4MasccAJQ&pid=ImgRaw&r=0&sres=1&sresct=1)

## **Propiedades y Características** 🔍

Las funciones compuestas poseen varias propiedades importantes:

- **Asociatividad**: Las funciones compuestas son asociativas, es decir, $(f \circ g) \circ h = f \circ (g \circ h)$.
- **Identidad**: Existe una función identidad $I(x) = x$ tal que $(f \circ I)(x) = f(x)$.

![Texto alternativo](https://th.bing.com/th/id/R.1ff5b905924ba52c6aaaf29b503d8713?rik=6W8UYtX42NfG0A&pid=ImgRaw&r=0)

## **Ejemplo de Función Compuesta** 📊

Supongamos las funciones $f(x) = 2x + 1$ y $g(x) = x^2$. La función compuesta $(f \circ g)(x)$ se calcula como:
$$(f \circ g)(x) = f(g(x)) = 2x^2 + 1$$

## **Aplicaciones y Significado** 🌐

Las funciones compuestas tienen aplicaciones en diversas áreas:
- **Física**: Modelan transformaciones en fenómenos físicos como la velocidad respecto al tiempo.
![Texto alternativo](https://th.bing.com/th/id/OIP.DExGtRVksdk5W6oUdDbwkAAAAA?rs=1&pid=ImgDetMain)


- **Economía**: Describen relaciones complejas en modelos económicos, como la oferta y la demanda.

![Texto alternativo](https://media.licdn.com/dms/image/C4D12AQEhwA22E_v9MA/article-cover_image-shrink_720_1280/0/1539303317794?e=2147483647&v=beta&t=pA7SACzLsI-910O-gGFGARn2uiTG47bathJAeIzFzDg)

- **Informática**: Fundamentales en programación para diseñar algoritmos eficientes y estructurar datos.

![Texto alternativo](https://www.tuyu.es/wp-content/uploads/2017/07/security23.jpg)

## **Importancia en Análisis Matemático** 📈

En el análisis matemático, las funciones compuestas son fundamentales para:
- **Simplificación de Expresiones**: Permite reducir expresiones complejas a formas más manejables.
- **Resolución de Problemas**: Facilita la solución de problemas relacionados con la variación de variables y funciones.

## **Conclusión** 🌟

Comprender a fondo las funciones compuestas no solo es esencial para la matemática avanzada, sino también para su aplicación práctica en disciplinas como la física, economía e informática. Su versatilidad y capacidad para modelar sistemas dinámicos y relaciones complejas las convierten en una herramienta poderosa en el análisis y la resolución de problemas reales.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de las funciones individuales g(x) y f(x)
def g(x):
    return x**2  # Función g(x) = x^2

def f(x):
    return np.sqrt(x)  # Función f(x) = sqrt(x)

# Definición de la función compuesta f(g(x))
def composite_function(x):
    return f(g(x))  # Función compuesta f(g(x)) = f(g(x))

# Crear un rango de valores de x para graficar
x = np.linspace(0, 3, 400)

# Calcular los valores de la función compuesta para cada valor de x
y = composite_function(x)

# Graficar la función compuesta
plt.figure(figsize=(8, 6))  # Tamaño del gráfico
plt.plot(x, y, label='$f(g(x)) = \sqrt{x^2}$', color='b')  # Graficar f(g(x)), etiqueta en LaTeX
plt.title('Función Compuesta $f(g(x))$')  # Título del gráfico
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('f(g(x))')  # Etiqueta del eje y
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar cuadrícula en el gráfico
plt.show()  # Mostrar el gráfico
```

# **Funcion de Utilidad** 📊🔍

## **Introducción** 📘
La función de utilidad es una herramienta crucial en economía y otras ciencias sociales para modelar y entender las preferencias de los individuos. Esta función permite cuantificar el grado de satisfacción o bienestar que un individuo obtiene de diferentes bienes o servicios.

## **Concepto de Utilidad** 🎯
La **utilidad** es una medida de la satisfacción o felicidad que un consumidor obtiene al consumir bienes y servicios. En términos matemáticos, una función de utilidad $ U(x) $ asigna un valor numérico a cada posible cesta de consumo $ x $, reflejando el nivel de satisfacción del consumidor.

## **Tipos de Funciones de Utilidad** 📏

1. **Función de Utilidad Cardinal** 🧮:
   - Asigna valores numéricos a diferentes niveles de satisfacción.
   - Ejemplo: $ U(x) = 3x_1 + 5x_2 $, donde $ x_1 $ y $ x_2 $ son cantidades de bienes consumidos.

2. **Función de Utilidad Ordinal** 📊:
   - Solo importa el orden de las preferencias, no la magnitud.
   - Ejemplo: Si $ U(A) > U(B) $, el consumidor prefiere A sobre B, pero no se sabe cuánto más.

## **Propiedades de las Funciones de Utilidad** 🌟

- **Monotonicidad** 📈: Si un consumidor prefiere más de un bien, entonces la función de utilidad debe ser no decreciente en ese bien.
- **Convexidad** 🌀: Refleja la preferencia por la diversidad. Las combinaciones de bienes son preferidas a los extremos.
- **Continuidad** 🔄: Las preferencias no cambian abruptamente con pequeños cambios en las cantidades consumidas.

## **Aplicaciones en Economía** 📚

1. **Teoría del Consumidor** 🛒:
   - **Equilibrio del Consumidor** ⚖️: El consumidor maximiza su utilidad sujeta a su restricción presupuestaria. Matemáticamente, esto se expresa como:

     $$
     \max U(x_1, x_2) \\
     \text{sujeto a} \, p_1x_1 + p_2x_2 \leq M
     $$

     Donde $ p_1 $ y $ p_2 $ son los precios de los bienes y $ M $ es el ingreso del consumidor.

2. **Curvas de Indiferencia** 🏞️:
   - Representan combinaciones de bienes que proporcionan el mismo nivel de utilidad al consumidor. Las curvas más altas representan mayor utilidad.

3. **Elasticidad de Sustitución** 🔄:
   - Mide la facilidad con la que un bien puede ser sustituido por otro mientras se mantiene el mismo nivel de utilidad.

### **Aplicaciones en Ciencias Políticas** 🗳️

1. **Elección Racional** 🧠:
   - Los votantes eligen candidatos o políticas que maximizan su utilidad esperada.
   - Modelo de utilidad esperada: $ U_i(C) = \sum_{j=1}^{n} p_j u_{ij} $
     - Donde $ p_j $ es la probabilidad percibida del resultado $ j $ y $ u_{ij} $ es la utilidad del individuo $ i $ para el resultado $ j $.

   - Ejemplo: En un juego de negociación, cada jugador maximiza su función de utilidad sujeta a las estrategias de los otros jugadores.

#### **Conclusión** 🎓
La función de utilidad es un concepto fundamental en las ciencias sociales, especialmente en economía, para modelar las preferencias y el comportamiento de los individuos. Su aplicación abarca desde la teoría del consumidor y la elección racional hasta la evaluación de políticas públicas y el análisis de bienestar social.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de utilidad Cobb-Douglas
def utilidad(x, y, alpha=0.5):
    return (x**alpha) * (y**(1-alpha))

# Crear valores de X e Y para graficar las curvas de indiferencia
x = np.linspace(0.1, 10, 400)
y = np.linspace(0.1, 10, 400)
X, Y = np.meshgrid(x, y)

# Calcular la utilidad para diferentes combinaciones de X e Y
U = utilidad(X, Y)

# Definir niveles de utilidad para las curvas de indiferencia
niveles_utilidad = [1, 2, 3, 4, 5]

# Crear el gráfico
plt.figure(figsize=(10, 6))

# Graficar las curvas de indiferencia
contornos = plt.contour(X, Y, U, levels=niveles_utilidad, linewidths=2, alpha=0.75, linestyles='dashed')
plt.clabel(contornos, inline=1, fontsize=10)

# Añadir etiquetas y título
plt.xlabel('Bien X')
plt.ylabel('Bien Y')
plt.title('Curvas de Indiferencia para una Función de Utilidad Cobb-Douglas')
plt.colorbar(label='Utilidad')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
