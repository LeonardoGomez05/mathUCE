# **Ecuaci贸n Punto-Pendiente**

La ecuaci贸n punto-pendiente es una forma 煤til para encontrar la ecuaci贸n de una recta cuando conoces un punto en la recta y su pendiente. La ecuaci贸n punto-pendiente de una l铆nea recta se representa como:

   $m = \frac{7 - 3}{4 - 2} = \frac{4}{2} = 2$

## **Funciones Exponenciales:**
Las funciones exponenciales son de la forma $f(x) = a^x$, donde $a > 0$ y $a \neq 1$. Estas funciones se caracterizan por un crecimiento o decaimiento r谩pido y son fundamentales en muchos campos como la biolog铆a, econom铆a, f铆sica, y m谩s.

- **Dominio**: Todos los n煤meros reales $x$.
- **Recorrido**:
  - Si $a > 1$: $(0, +\infty)$.
  - Si $0 < a < 1$: $(0, +\infty)$.
- **As铆ntota Horizontal**: La recta $y = 0$.
- **Intersecci贸n con el Eje y**: $(0, 1)$.
- **Monoton铆a**:
  - Creciente si $a > 1$. La funci贸n aumenta a medida que $x$ aumenta.
  - Decreciente si $0 < a < 1$. La funci贸n disminuye a medida que $x$ aumenta.
- **Convexidad**: Siempre c贸ncava hacia arriba. Esto significa que su segunda derivada es positiva.
- **Comportamiento Asint贸tico**:
  - Si $a > 1$, a medida que $x \rightarrow -\infty$, $a^x \rightarrow 0$ y cuando $x \rightarrow \infty$, $a^x \rightarrow \infty$.
  - Si $0 < a < 1$, a medida que $x \rightarrow -\infty$, $a^x \rightarrow \infty$ y cuando $x \rightarrow \infty$, $a^x \rightarrow 0$.
- **Aplicaciones**: Modelado de crecimiento poblacional , desintegraci贸n radiactiva 锔, inter茅s compuesto, entre otros.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n exponencial
def exponential_function(x, a):
    return a ** x

# Datos de ejemplo
x = np.linspace(-2, 2, 400)
a = 2  # Cambia el valor de 'a' seg煤n necesites

# Calcular los valores de y
y = exponential_function(x, a)

# Graficar la funci贸n exponencial
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'$f(x) = {a}^x$')
plt.title('Funci贸n Exponencial')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## **Funciones Logar铆tmicas**
Las funciones logar铆tmicas son inversas de las funciones exponenciales y tienen la forma $f(x) = \log_a(x)$, donde $a$ es la base del logaritmo y $a > 0$ y $a \neq 1$.

  - Creciente si $a > 1$. La funci贸n aumenta a medida que $x$ aumenta.

- **Convexidad**:
  - C贸ncava hacia arriba si $a > 1$. Esto significa que su segunda derivada es negativa.
  - C贸ncava hacia abajo si $0 < a < 1$. Esto significa que su segunda derivada es positiva.
- **Comportamiento Asint贸tico**: A medida que $x \rightarrow 0$, $\log_a(x) \rightarrow -\infty$, y cuando $x \rightarrow \infty$, $\log_a(x) \rightarrow \infty$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n logar铆tmica
def logarithmic_function(x, a):
    return np.log(x) / np.log(a)

# Datos de ejemplo
x = np.linspace(0.01, 5, 400)
a = 2  # Cambia el valor de 'a' seg煤n necesites

# Calcular los valores de y
y = logarithmic_function(x, a)

# Graficar la funci贸n logar铆tmica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'$f(x) = \log_{a}(x)$')
plt.title('Funci贸n Logar铆tmica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## **Funciones Racionales**
Las funciones racionales son cocientes de polinomios y tienen la forma $f(x) = \frac{p(x)}{q(x)}$, donde $p(x)$ y $q(x)$ son polinomios y $q(x) \neq 0$.

- **Dominio**: Todos los valores de $x$ excepto donde $q(x) = 0$.
- **Recorrido**: Determinado por el comportamiento asint贸tico y los valores de $p(x)$ y $q(x)$.
- **As铆ntotas Verticales**: En los puntos donde $q(x) = 0$.
- **As铆ntotas Horizontales**: Determinadas por el grado relativo de $p(x)$ y $q(x)$.
  - Si el grado de $p(x)$ es menor que el grado de $q(x)$, la as铆ntota horizontal es $y = 0$.
  - Si el grado de $p(x)$ es igual al grado de $q(x)$, la as铆ntota horizontal es $y = \frac{a_n}{b_n}$, donde $a_n$ y $b_n$ son los coeficientes l铆deres de $p(x)$ y $q(x)$, respectivamente.
  - Si el grado de $p(x)$ es mayor que el grado de $q(x)$, no hay as铆ntota horizontal; en su lugar, hay una as铆ntota oblicua.
- **Intersecciones con los Ejes**: Dependientes de los valores de $p(x)$ y $q(x)$ en esos puntos.
  - Intersecci贸n con el eje $y$: Se encuentra evaluando $f(0)$, siempre que $q(0) \neq 0$.
  - Intersecci贸n con el eje $x$: Se encuentra resolviendo $p(x) = 0$.
- **Comportamiento Asint贸tico**: Var铆a seg煤n los grados de $p(x)$ y $q(x)$.
- **Aplicaciones**: Modelado de fen贸menos f铆sicos y econ贸micos que involucran relaciones proporcionales o fraccionales , an谩lisis de circuitos el茅ctricos , din谩mica de poblaciones , entre otros.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n racional
def rational_function(x, p, q):
    return p(x) / q(x)

# Polinomios de ejemplo
p = lambda x: 2*x + 1
q = lambda x: x**2 + 1

# Datos de ejemplo
x = np.linspace(-5, 5, 400)

# Calcular los valores de y
y = rational_function(x, p, q)

# Graficar la funci贸n racional
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'$f(x) = \\frac{{2x + 1}}{{x^2 + 1}}$')
plt.title('Funci贸n Racional')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## **Funciones Trigonom茅tricas**
Las funciones trigonom茅tricas incluyen seno, coseno, tangente, cotangente, secante y cosecante. Son funciones peri贸dicas que describen relaciones angulares y longitudes en tri谩ngulos.

- **Tipos**:
  - Seno: $f(x) = \sin(x)$
  - Coseno: $f(x) = \cos(x)$
  - Tangente: $f(x) = \tan(x)$
  - Cotangente: $f(x) = \cot(x)$
  - Secante: $f(x) = \sec(x)$
  - Cosecante: $f(x) = \csc(x)$
- **Dominio**: Todos los n煤meros reales, con exclusiones para tangente, cotangente, secante y cosecante en los puntos donde estas funciones no est谩n definidas.
  - $\sin(x)$ y $\cos(x)$: Todos los n煤meros reales.
  - $\tan(x)$ y $\sec(x)$: Todos los n煤meros reales excepto $x = \frac{\pi}{2} + k\pi$.
  - $\cot(x)$ y $\csc(x)$: Todos los n煤meros reales excepto $x = k\pi$.
- **Recorrido**:
  - Seno y Coseno: $[-1, 1]$.
  - Tangente y Cotangente: Todos los n煤meros reales.
  - Secante y Cosecante: $(-\infty, -1] \cup [1, +\infty)$.
- **Periodicidad**: Repetici贸n regular de valores a intervalos espec铆ficos.
  - $\sin(x)$ y $\cos(x)$: Per铆odo $2\pi$.
  - $\tan(x)$ y $\cot(x)$: Per铆odo $\pi$.
  - $\sec(x)$ y $\csc(x)$: Per铆odo $2\pi$.
- **Intersecciones con los Ejes**: Dependientes de la funci贸n espec铆fica y su periodicidad.
  - $\sin(x)$: Intersecci贸n en $(k\pi, 0)$.
  - $\cos(x)$: Intersecci贸n en $(\frac{\pi}{2} + k\pi, 0)$.
  - $\tan(x)$: Intersecci贸n en $(k\pi, 0)$.
- **Comportamiento Asint贸tico**:
  - $\sin(x)$ y $\cos(x)$ no tienen as铆ntotas.
  - $\tan(x)$ y $\sec(x)$ tienen as铆ntotas verticales en $x = \frac{\pi}{2} + k\pi$.
  - $\cot(x)$ y $\csc(x)$ tienen as铆ntotas verticales en $x = k\pi$.
- **Aplicaciones**: Modelado de fen贸menos peri贸dicos como movimiento oscilatorio , ondas y vibraciones , an谩lisis de se帽ales , geometr铆a de tri谩ngulos , astronom铆a , entre otros.




```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n trigonom茅trica
def trigonometric_function(x, type='sin'):
    if type == 'sin':
        return np.sin(x)
    elif type == 'cos':
        return np.cos(x)
    elif type == 'tan':
        return np.tan(x)
    elif type == 'cot':
        return 1 / np.tan(x)
    elif type == 'sec':
        return 1 / np.cos(x)
    elif type == 'csc':
        return 1 / np.sin(x)
    else:
        raise ValueError('Tipo de funci贸n trigonom茅trica no v谩lido')

# Datos de ejemplo
x = np.linspace(-2*np.pi, 2*np.pi, 400)

# Calcular los valores de y (por ejemplo, seno)
y = trigonometric_function(x, type='sin')

# Graficar la funci贸n trigonom茅trica (por ejemplo, seno)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = \sin(x)$')
plt.title('Funci贸n Trigonom茅trica (Seno)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

### **Funciones Polin贸micas**
Las funciones polin贸micas son sumas de potencias de $x$ con coeficientes constantes y tienen la forma $f(x) = a_n x^n + a_{n-1} x^{n-1} + \ldots + a_1 x + a_0$, donde $a_i$ son coeficientes constantes y $n$ es el grado del polinomio.

- **Dominio**: Todos los n煤meros reales.
- **Recorrido**: Determinado por el grado y los coeficientes del polinomio. En general, $(-\infty, +\infty)$.
- **Intersecci贸n con el Eje y**: $(0, a_0)$.
- **Ra铆ces**: Los valores de $x$ donde $f(x) = 0$. Pueden ser reales o complejas.
- **Crecimiento**:
  - Polinomios de grado par: Crecen hacia $+\infty$ en ambos extremos o decrecen hacia $-\infty$ en ambos extremos.
  - Polinomios de grado impar: Crecen hacia $+\infty$ en un extremo y decrecen hacia $-\infty$ en el otro.
- **Comportamiento Asint贸tico**: Determinado por el t茅rmino de mayor grado. La funci贸n se comporta como $a_n x^n$ para valores grandes de $|x|$.
- **Monoton铆a y Extremos**:
  - Los puntos cr铆ticos se encuentran derivando la funci贸n y resolviendo $f'(x) = 0$.
  - La naturaleza de estos puntos (m谩ximos, m铆nimos o puntos de inflexi贸n) se determina con la segunda derivada.
- **Convexidad**:
  - C贸ncava hacia arriba (convexa) donde la segunda derivada $f''(x) > 0$.
  - C贸ncava hacia abajo (c贸ncava) donde la segunda derivada $f''(x) < 0$.
- **Aplicaciones**: Ampliamente utilizado en modelado matem谩tico de fen贸menos f铆sicos , econ贸micos  y sociales , resoluci贸n de ecuaciones algebraicas , interpolaci贸n y aproximaci贸n de funciones , entre otros.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n polin贸mica
def polynomial_function(x, coeffs):
    return np.polyval(coeffs, x)

# Coeficientes del polinomio de ejemplo: f(x) = 2x^3 - x^2 + 3x - 5
coeffs = [2, -1, 3, -5]

# Datos de ejemplo
x = np.linspace(-2, 2, 400)

# Calcular los valores de y
y = polynomial_function(x, coeffs)

# Graficar la funci贸n polin贸mica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = 2x^3 - x^2 + 3x - 5$')
plt.title('Funci贸n Polin贸mica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

- $\alpha$ y $\beta$ son par谩metros que representan las elasticidades de la producci贸n respecto al trabajo y al capital, respectivamente, y $A$ es una constante de escala.

Esta funci贸n permite a las empresas determinar c贸mo ajustar sus niveles de insumos para maximizar la producci贸n dada una restricci贸n de costos.

Al comprender sus propiedades avanzadas y aplicaciones pr谩cticas, podemos analizar de manera m谩s rigurosa la asignaci贸n de recursos, la eficiencia productiva y el bienestar econ贸mico en diferentes contextos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Funci贸n de utilidad cuadr谩tica
def utilidad_cuadratica(x1, x2):
    return 3 * x1**2 + 2 * x2**2

# Definir rangos de variables
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(0, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
Z_utilidad = utilidad_cuadratica(X1, X2)

# Graficar funci贸n de utilidad
plt.figure(figsize=(10, 6))
plt.contour(X1, X2, Z_utilidad, levels=20, cmap='viridis')
plt.colorbar(label='Utilidad')
plt.xlabel('Cantidad de Bien 1')
plt.ylabel('Cantidad de Bien 2')
plt.title('Funci贸n de Utilidad Cuadr谩tica')
plt.grid(True)
plt.show()

# Funci贸n de producci贸n Cobb-Douglas
def produccion_cobb_douglas(L, K):
    alpha = 0.3
    beta = 0.7
    A = 10
    return A * (L**alpha) * (K**beta)

# Definir rangos de insumos
L = np.linspace(0, 100, 100)
K = np.linspace(0, 100, 100)
L, K = np.meshgrid(L, K)
Z_produccion = produccion_cobb_douglas(L, K)

# Graficar funci贸n de producci贸n
plt.figure(figsize=(10, 6))
plt.contour(L, K, Z_produccion, levels=20, cmap='plasma')
plt.colorbar(label='Producci贸n')
plt.xlabel('Trabajo (L)')
plt.ylabel('Capital (K)')
plt.title('Funci贸n de Producci贸n Cobb-Douglas')
plt.grid(True)
plt.show()
```

**Funci贸n de Utilidad Cuadr谩tica:** Muestra curvas de nivel que representan diferentes combinaciones de bienes que proporcionan la misma utilidad al consumidor.

**Funci贸n de Producci贸n Cobb-Douglas:** Muestra curvas de nivel que representan diferentes combinaciones de trabajo y capital que producen la misma cantidad de producci贸n.
