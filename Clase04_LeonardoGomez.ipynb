# **Historia de las Funciones** 📚

El origen de las funciones es bastante peculiar, ya que no es como una regla científica o una fórmula que se descubre en un año determinado, sino que han tenido que pasar muchísimos años para asimilar este concepto.

Cabe destacar que en las civilizaciones antiguas de Mesopotamia, Egipto y Grecia ya se utilizaban algunas nociones de las funciones. Sin embargo, este no se considera como el inicio de las funciones porque muy probablemente aún no entendían el concepto abstracto de función, es decir, no deberían ser del todo conscientes que utilizaban funciones para hacer cálculos. 🏺🧮

Aunque no por unanimidad, la mayor parte de historiadores de las matemáticas atribuyen el inicio de las funciones al científico Nicole Oresme (1323-1382), cuando dijo que las leyes de la naturaleza son como relaciones de dependencia entre dos magnitudes. 📊

Más adelante, el astrónomo Galileo Galilei (1564-1642) realizó estudios sobre el movimiento con los que es evidente que ya se entendía el significado de una función matemática. Entre otros ejemplos, demostró que cada número natural tiene un único cuadrado perfecto y que, por lo tanto, existe una relación matemática entre los números naturales y los cuadrados perfectos. 🌠

El término función no aparece hasta finales del siglo XVII, cuando René Descartes (1596-1650), Isaac Newton (1643-1727) y Gottfried Leibniz (1646-1716) definieron una función como una dependencia entre dos cantidades variables. Pero la notación $f(x)$ fue introducida un poco más tarde por Alexis Claude Clairaut (1713-1765) y Leonhard Euler (1707-1783). 🔢🧠

A pesar de que en ese momento ya se utilizaban las funciones para hacer cálculos algebraicos, e incluso se publicaban teorías y libros sobre las funciones, su definición aún no estaba completa y pasaron siglos hasta que el matemático Édouard Goursat (1858-1936) en el año 1923 llegó a la definición de función que en la mayoría de libros actuales se usa: «se dice que y es una función de x si a cada valor de x le corresponde un único valor de y. Esta correspondencia se indica mediante la ecuación $y = f(x)$». 📘

## 💹**Funciones**

### **Definición de Función**
Una **función** es una relación matemática entre dos conjuntos, en la cual a cada elemento del primer conjunto $dominio$ le corresponde un único elemento del segundo conjunto $codominio$. Formalmente, una función $f$ de un conjunto $A$ a un conjunto $B$ se denota como $f: A \to B$.

### **Imagen de una Función**
La **imagen** (o rango) de una función es el conjunto de todos los valores de salida que puede tomar la función. Es decir, si $f: A \to B$ es una función, entonces la imagen de $f$ es el subconjunto de $B$ definido como:
$\text{Im}(f) = \{ f(x) \mid x \in A \}$

#### **Ejemplo:**
Consideremos la función $f: \mathbb{R} \to \mathbb{R}$ definida por $f(x) = x^2$. En este caso, la imagen de $f$ es el conjunto de todos los números reales no negativos:
$\text{Im}(f) = \{ y \in \mathbb{R} \mid y \geq 0 \}$

### **Preimagen de una Función**
La **preimagen** de un subconjunto $C$ del codominio es el conjunto de todos los elementos del dominio que se mapean en los elementos de $C$ bajo la función $f$. Formalmente, para un subconjunto $C \subseteq B$, la preimagen de $C$ bajo $f$ se define como:
$f^{-1}(C) = \{ x \in A \mid f(x) \in C \}$

#### **Ejemplo:**
Usando la misma función $f(x) = x^2$, consideremos el subconjunto $C = \{4, 9\}$. La preimagen de $C$ es:
$f^{-1}(C) = \{ x \in \mathbb{R} \mid f(x) \in \{4, 9\} \} = \{ -3, -2, 2, 3 \}$


# **Tipos de funciones**
## 🔍**Función Lineal**

Una función lineal es una función matemática que puede representarse mediante una ecuación de la forma:
$f(x) = mx$ donde:

### **Variables:**
- **Variable independiente o exógena ($x$):** Es el valor de entrada para la función. Puede tomar cualquier valor dentro del dominio de la función.
- **Variable dependiente o endógena ($y$) o $f(x)$:** Es el valor de salida de la función, que depende de la variable independiente ($x$).

### **Parámetros:**
- **Pendiente ($m$):** Es el coeficiente que determina la inclinación de la recta. Es un valor que se fija para definir una función específica y puede variar entre diferentes funciones lineales.

### **Constantes:**
- En una función lineal, no hay una constante aparte de la pendiente ($m$) cuando se refiere a la ecuación $f(x) = mx$. Sin embargo, si consideramos la función en su forma general, los únicos términos que pueden considerarse constantes son los que no dependen de $x$. En este caso, solo la pendiente ($m$) actúa como un valor fijo, aunque no es una constante en el sentido estricto de una función constante.


**Características de la función lineal:**

1. **Pendiente ($m$):** Indica la inclinación de la recta y representa el cambio en $y$ por unidad de cambio en $x$. Si $m$ es positiva, la recta sube al avanzar en el eje $x$. Si $m$ es negativa, la recta baja al avanzar en el eje $x$.
2. **Recta que pasa por el origen:** Dado que no hay término constante $b$, todas las funciones lineales pasan por el origen $(0,0)$.

**Ejemplo:**
$f(x) = 2x$
Esta función tiene una pendiente de 2, lo que significa que por cada unidad que $x$ aumenta, $y$ aumenta en 2 unidades.

- **Variable independiente:** $x$
- **Variable dependiente:** $f(x)$ o $y$
- **Parámetro:** $m = 2$
- No hay constante aparte del parámetro pendiente.

## 🔍**Función Afín**

Una función afín es una función que puede representarse mediante una ecuación de la forma:
$f(x) = mx + b$ donde:

**Variables:**
- **Variable independiente ($x$):** Es el valor de entrada para la función. Puede tomar cualquier valor dentro del dominio de la función.
- **Variable dependiente ($y$) o $f(x)$:** Es el valor de salida de la función, que depende de la variable independiente ($x$).

**Parámetros:**
- **Pendiente ($m$):** Es el coeficiente que determina la inclinación de la recta. Al igual que en la función lineal, es un valor que se fija para definir una función específica.
- **Intersección \( b \):** Es el término independiente que determina el punto donde la recta cruza el eje ($y$). También se fija para definir una función específica.

**Constantes:**
- **($b$):** El término independiente ($b$) es una constante porque no cambia con el valor de ($x$). En la ecuación, ($b$) actúa como una constante que desplaza la recta verticalmente.

**Características de la función afín:**

1. **Pendiente ($m$):** Similar a la función lineal, $m$ indica la inclinación de la recta.
2. **Intersección con el eje y ($b$):** El término $b$ desplaza la recta verticalmente, indicando dónde la recta cruza el eje $y$.

**Ejemplo:**
$f(x) = 2x + 3$
Esta función tiene una pendiente de 2 y una intersección con el eje $y$ en 3. Esto significa que por cada unidad que $x$ aumenta, $y$ aumenta en 2 unidades, y la recta cruza el eje $y$ en el punto $(0, 3)$.

- **Variable independiente:** $x$
- **Variable dependiente:** $f(x)$ o $y$
- **Parámetro:** $m = 2$
- **Constante:** $b = 3$

## **Representación Gráfica**

- **Función Lineal:** La gráfica de una función lineal es una recta que pasa por el origen $(0,0)$ con pendiente $m$.
- **Función Afín:** La gráfica de una función afín es una recta que cruza el eje $y$ en el punto ($0$ , $b$) y tiene pendiente $m$.

## **Ejemplos en Python**



```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos la función lineal
def funcion_lineal(x):
    return 2 * x

# Definimos la función afín
def funcion_afin(x):
    return 2 * x + 3

# Generamos valores de x
x = np.linspace(-10, 10, 400)

# Calculamos los valores de y para ambas funciones
y_lineal = funcion_lineal(x)
y_afin = funcion_afin(x)

# Creamos la gráfica
plt.figure(figsize=(10, 6))

# Graficamos la función lineal
plt.plot(x, y_lineal, label='Función Lineal: $y = 2x$', color='blue')

# Graficamos la función afín
plt.plot(x, y_afin, label='Función Afín: $y = 2x + 3$', color='red')

# Añadimos etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráficas de Función Lineal y Función Afín')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Mostramos la gráfica
plt.show()
```

## **Tipos de pendiente (m)**
Una función de primer grado puede tener diferentes tipos de pendiente, que dependen del coeficiente de la variable independiente ($x$). Aquí están los tipos más comunes:

1. **Pendiente positiva ($m>0$)**: Cuando el coeficiente de x (el número delante de x) es positivo, la función tiene una pendiente ascendente. Esto significa que a medida que aumenta x, también aumenta y.

2. **Pendiente negativa ($m<0$)**: Si el coeficiente de x es negativo, la función tiene una pendiente descendente. En este caso, a medida que x aumenta, y disminuye.

3. **Pendiente cero ($m=0$)**: Cuando el coeficiente de x es cero, la función es una línea horizontal. Esto significa que no hay cambio en y respecto a cambios en x; y es constante.

4. **Pendiente indefinida o no definida ($m=∞$)**: Esto ocurre cuando el coeficiente de x es cero, pero hay un término constante en la función. En este caso, la función es una línea vertical, lo cual no es una función de primer grado típica, pero es un caso especial que debe mencionarse.

# **Ecuación Punto-Pendiente**

La ecuación punto-pendiente es una forma útil para encontrar la ecuación de una recta cuando conoces un punto en la recta y su pendiente. La ecuación punto-pendiente de una línea recta se representa como:

$y - y_1 = m(x - x_1)$

donde:
- $(x_1, y_1)$ es un punto en la recta.
- \$m$ es la pendiente de la recta.

### Pasos para Encontrar la Ecuación de la Recta Usando la Ecuación Punto-Pendiente

1. **Calcular la Pendiente ($m$)**:
   Si tienes dos puntos $(x_1, y_1)$ y $(x_2, y_2)$, la pendiente $m$ se calcula como:
   $m = \frac{y_2 - y_1}{x_2 - x_1}$

2. **Usar la Ecuación Punto-Pendiente**:
   Con la pendiente $m$ y uno de los puntos $(x_1, y_1)$, sustituyes estos valores en la ecuación punto-pendiente.

### **Ejemplo Específico**

Supongamos que tenemos dos puntos $(2, 3)$ y $(4, 7)$:

1. **Calcular la Pendiente ($m$)**:
   $m = \frac{7 - 3}{4 - 2} = \frac{4}{2} = 2$

2. **Usar la Ecuación Punto-Pendiente**:
   Usando el punto $(2, 3)$ y la pendiente $m = 2$:

   $y - 3 = 2(x - 2)$

   Simplificando la ecuación:

   $y - 3 = 2x - 4$

   $y = 2x - 1$

Por lo tanto, la ecuación de la recta que pasa por los puntos $(2, 3)$ y $(4, 7)$ es:

$y = 2x - 1$

# 🔍**Función Cuadrática**

La función cuadrática es una función polinómica de segundo grado, generalmente escrita en la forma:

$f(x) = ax^2 + bx + c$

donde ($a$), ($b$), y ($c$) son coeficientes constantes y  $a \neq 0$ para que la función sea cuadrática (es decir, el término cuadrático no puede ser cero).

### **Características principales de la función cuadrática:**

1. **Forma estándar y coeficientes:**
   - **Forma estándar:**

   $f(x) = ax^2 + bx + c$
   - ($a$): coeficiente cuadrático.
   - ($b$): coeficiente lineal.
   - ($c$): término constante.

2. **Gráfico:**
   - La gráfica de una función cuadrática es una parábola.
   - Si $( a > 0 )$, la parábola se abre hacia arriba.
   - Si $( a < 0 )$, la parábola se abre hacia abajo.

![Texto alternativo](https://th.bing.com/th/id/R.864e1e62c0996fd11a85ab39dc571420?rik=Xqa4g4R5bwpN1w&riu=http%3a%2f%2fimage.slidesharecdn.com%2fintroduccion-130329191704-phpapp02%2f95%2fintroduccion-12-638.jpg%3fcb%3d1364585081&ehk=yzPwRrD7PKuKdZv%2bo%2ffflsm%2f04TMZO1ChORbg7JFiEI%3d&risl=&pid=ImgRaw&r=0&sres=1&sresct=1)


3. **Vértice de la parábola:**
   - El vértice de la parábola $f(x) = ax^2 + bx + c$ se encuentra en $x = -\frac{b}{2a}$.
   - La coordenada $y$ del vértice se obtiene sustituyendo $x = -\frac{b}{2a}$ en la función.

![Texto alternativo](https://th.bing.com/th/id/OIP.QoWxZOYeAB-bjpcQ7lYrxQHaH0?rs=1&pid=ImgDetMain)

4. **Eje de simetría:**
   - El eje de simetría de la parábola es la línea vertical que pasa por el vértice.
   - La ecuación del eje de simetría es $x = -\frac{b}{2a} \$.

5. **Raíces o soluciones:**
   - Las raíces de la función cuadrática son los valores de $(x)$ para los cuales $f(x) = 0$.
   - Las raíces se calculan usando la fórmula cuadrática: $x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$.
   - Dependiendo del discriminante $\Delta = b^2 - 4ac$:
     - Si $\Delta > 0$, la función tiene dos raíces reales distintas.
     - Si $\Delta = 0$, la función tiene una raíz real doble (la parábola toca el eje \( x \) en un punto).
     - Si $\Delta < 0$, la función tiene raíces complejas conjugadas (no hay raíces reales).

![Texto alternativo](https://www.neurochispas.com/wp-content/uploads/2022/07/Graficas-de-ecuaciones-cuadraticas-con-diferente-discriminante-1024x378.png)

6. **Intersecciones con los ejes:**
   - **Intersección con el eje $(y)$**: Se obtiene sustituyendo $x = 0$ en la función $f(x)$, es decir, $y = f(0) = c$.
   - **Intersección con el eje $(x)$**: Se encuentran las raíces de la función usando la fórmula cuadrática.

### **Dominio y Rango de una Función Cuadrática**

El dominio de una función cuadrática $f(x) = ax^2 + bx + c$ es el conjunto de todos los valores de $x$ para los cuales la función está definida. En el caso de una función polinómica como la cuadrática, su dominio son todos los números reales.

El rango de una función cuadrática depende de la dirección en la que abre la parábola

#### Ejemplo:

Considera la función cuadrática $f(x) = x^2 - 4x + 3$:

1. **Dominio:** Como es una función polinómica, su dominio es $\mathbb{R}$.

2. **Rango:**
   - El vértice de la parábola se encuentra en $x = -\frac{-4}{2 \cdot 1} = 2$.
   - Evaluando en $x = 2$: $f(2) = 2^2 - 4 \cdot 2 + 3 = -1$.
   - Dado que $a > 0$, la parábola abre hacia arriba, por lo tanto, el rango es $[-1, \infty)$.



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función cuadrática
def funcion_cuadratica(x):
    return x**2 - 4*x + 3

# Generar valores de x
x = np.linspace(-1, 5, 400)  # Valores de x desde -1 hasta 5

# Calcular los valores de y para la función cuadrática
y = funcion_cuadratica(x)

# Crear la figura y los ejes
plt.figure(figsize=(8, 6))  # Tamaño de la figura
plt.plot(x, y, label='$f(x) = x^2 - 4x + 3$', color='blue', linewidth=2)  # Graficar la función cuadrática
plt.axhline(0, color='black', linewidth=0.5)  # Eje x
plt.axvline(0, color='black', linewidth=0.5)  # Eje y

# Configuración de etiquetas y título
plt.title('Gráfica de la Función Cuadrática $f(x) = x^2 - 4x + 3$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Ajustes adicionales
plt.grid(True)  # Cuadrícula en la gráfica
plt.ylim(-2, 6)  # Rango vertical de la gráfica
plt.xlim(-1, 5)  # Rango horizontal de la gráfica

# Mostrar la gráfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def funcion_cuadratica(x, a, b, c):
    return a * x**2 + b * x + c

# Definir los coeficientes de la función cuadrática
a = 1
b = -3
c = 2

# Generar valores de x
x = np.linspace(-5, 5, 400)  # Valores de x desde -5 hasta 5

# Calcular los valores de y para la función cuadrática
y = funcion_cuadratica(x, a, b, c)

# Crear la figura y los ejes
plt.figure(figsize=(8, 6))  # Tamaño de la figura
plt.plot(x, y, label=f'{a}x^2 + {b}x + {c}', color='blue', linewidth=2)  # Graficar la función cuadrática
plt.axhline(0, color='black', linewidth=0.5)  # Eje x
plt.axvline(0, color='black', linewidth=0.5)  # Eje y

# Configuración de etiquetas y título
plt.title('Gráfica de una Función Cuadrática')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Ajustes adicionales
plt.grid(True)  # Cuadrícula en la gráfica
plt.ylim(-2, 6)  # Rango vertical de la gráfica
plt.xlim(-3, 4)  # Rango horizontal de la gráfica

# Mostrar la gráfica
plt.show()
```

```python
import math

def formula_general(a, b, c):
    # Calcular el discriminante
    discriminante = b**2 - 4 * a * c

    # Calcular las raíces
    if discriminante > 0:
        raiz1 = (-b + math.sqrt(discriminante)) / (2 * a)
        raiz2 = (-b - math.sqrt(discriminante)) / (2 * a)
        return raiz1, raiz2
    elif discriminante == 0:
        raiz = -b / (2 * a)
        return raiz
    else:
        raiz_real = -b / (2 * a)
        raiz_imaginaria = math.sqrt(abs(discriminante)) / (2 * a)
        return raiz_real, raiz_imaginaria

# Ejemplo de uso:
a = 1
b = -3
c = 2

raices = formula_general(a, b, c)
print("Las raíces son:", raices)
```

## 🔍**Funciones Exponenciales:**
Las funciones exponenciales son de la forma $f(x) = a^x$, donde $a > 0$ y $a \neq 1$. Estas funciones se caracterizan por un crecimiento o decaimiento rápido y son fundamentales en muchos campos como la biología, economía, física, y más.

- **Dominio**: Todos los números reales $x$.
- **Recorrido**:
  - Si $a > 1$: $(0, +\infty)$.
  - Si $0 < a < 1$: $(0, +\infty)$.
- **Asíntota Horizontal**: La recta $y = 0$.
- **Intersección con el Eje y**: $(0, 1)$.
- **Monotonía**:
  - Creciente si $a > 1$. La función aumenta a medida que $x$ aumenta.
  - Decreciente si $0 < a < 1$. La función disminuye a medida que $x$ aumenta.
- **Convexidad**: Siempre cóncava hacia arriba. Esto significa que su segunda derivada es positiva.
- **Comportamiento Asintótico**:
  - Si $a > 1$, a medida que $x \rightarrow -\infty$, $a^x \rightarrow 0$ y cuando $x \rightarrow \infty$, $a^x \rightarrow \infty$.
  - Si $0 < a < 1$, a medida que $x \rightarrow -\infty$, $a^x \rightarrow \infty$ y cuando $x \rightarrow \infty$, $a^x \rightarrow 0$.
- **Aplicaciones**: Modelado de crecimiento poblacional 🌱, desintegración radiactiva ☢️, interés compuesto, entre otros.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función exponencial
def exponential_function(x, a):
    return a ** x

# Datos de ejemplo
x = np.linspace(-2, 2, 400)
a = 2  # Cambia el valor de 'a' según necesites

# Calcular los valores de y
y = exponential_function(x, a)

# Graficar la función exponencial
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'$f(x) = {a}^x$')
plt.title('Función Exponencial')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## 🔍**Funciones Logarítmicas**
Las funciones logarítmicas son inversas de las funciones exponenciales y tienen la forma $f(x) = \log_a(x)$, donde $a$ es la base del logaritmo y $a > 0$ y $a \neq 1$.

- **Dominio**: $x > 0$ para bases reales positivas.
- **Recorrido**: Todos los números reales.
- **Asíntota Vertical**: La recta $x = 0$.
- **Intersección con el Eje x**: $(1, 0)$.
- **Monotonía**:
  - Creciente si $a > 1$. La función aumenta a medida que $x$ aumenta.
  - Decreciente si $0 < a < 1$. La función disminuye a medida que $x$ aumenta.
- **Convexidad**:
  - Cóncava hacia arriba si $a > 1$. Esto significa que su segunda derivada es negativa.
  - Cóncava hacia abajo si $0 < a < 1$. Esto significa que su segunda derivada es positiva.
- **Comportamiento Asintótico**: A medida que $x \rightarrow 0$, $\log_a(x) \rightarrow -\infty$, y cuando $x \rightarrow \infty$, $\log_a(x) \rightarrow \infty$.
- **Aplicaciones**: Resolución de ecuaciones exponenciales, modelado de fenómenos donde el crecimiento depende del tiempo ⏳, medición de intensidad sonora (decibelios) 🔊, escala de pH 🌊, entre otros.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función logarítmica
def logarithmic_function(x, a):
    return np.log(x) / np.log(a)

# Datos de ejemplo
x = np.linspace(0.01, 5, 400)
a = 2  # Cambia el valor de 'a' según necesites

# Calcular los valores de y
y = logarithmic_function(x, a)

# Graficar la función logarítmica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'$f(x) = \log_{a}(x)$')
plt.title('Función Logarítmica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## 🔍**Funciones Racionales**
Las funciones racionales son cocientes de polinomios y tienen la forma $f(x) = \frac{p(x)}{q(x)}$, donde $p(x)$ y $q(x)$ son polinomios y $q(x) \neq 0$.

- **Dominio**: Todos los valores de $x$ excepto donde $q(x) = 0$.
- **Recorrido**: Determinado por el comportamiento asintótico y los valores de $p(x)$ y $q(x)$.
- **Asíntotas Verticales**: En los puntos donde $q(x) = 0$.
- **Asíntotas Horizontales**: Determinadas por el grado relativo de $p(x)$ y $q(x)$.
  - Si el grado de $p(x)$ es menor que el grado de $q(x)$, la asíntota horizontal es $y = 0$.
  - Si el grado de $p(x)$ es igual al grado de $q(x)$, la asíntota horizontal es $y = \frac{a_n}{b_n}$, donde $a_n$ y $b_n$ son los coeficientes líderes de $p(x)$ y $q(x)$, respectivamente.
  - Si el grado de $p(x)$ es mayor que el grado de $q(x)$, no hay asíntota horizontal; en su lugar, hay una asíntota oblicua.
- **Intersecciones con los Ejes**: Dependientes de los valores de $p(x)$ y $q(x)$ en esos puntos.
  - Intersección con el eje $y$: Se encuentra evaluando $f(0)$, siempre que $q(0) \neq 0$.
  - Intersección con el eje $x$: Se encuentra resolviendo $p(x) = 0$.
- **Comportamiento Asintótico**: Varía según los grados de $p(x)$ y $q(x)$.
- **Aplicaciones**: Modelado de fenómenos físicos y económicos que involucran relaciones proporcionales o fraccionales 📊, análisis de circuitos eléctricos ⚡, dinámica de poblaciones 🌱, entre otros.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función racional
def rational_function(x, p, q):
    return p(x) / q(x)

# Polinomios de ejemplo
p = lambda x: 2*x + 1
q = lambda x: x**2 + 1

# Datos de ejemplo
x = np.linspace(-5, 5, 400)

# Calcular los valores de y
y = rational_function(x, p, q)

# Graficar la función racional
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'$f(x) = \\frac{{2x + 1}}{{x^2 + 1}}$')
plt.title('Función Racional')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## 🔍**Funciones Irracionales**
Las funciones irracionales involucran raíces de polinomios y tienen la forma general $f(x) = \sqrt[n]{g(x)}$, donde $g(x)$ es una función polinómica y $n$ es un entero positivo.

- **Dominio**: Valores de $x$ donde la expresión bajo la raíz es no negativa (para raíces pares) o todos los números reales (para raíces impares).
  - Para $\sqrt{x}$ (raíz cuadrada), el dominio es $x \geq 0$.
  - Para $\sqrt[3]{x}$ (raíz cúbica), el dominio es todos los números reales.
- **Recorrido**:
  - Para $n$ par: $[0, +\infty)$.
  - Para $n$ impar: Todos los números reales.
- **Intersección con el Eje y**: Dependiente de $g(x)$.
  - Para $\sqrt{x}$, la intersección es $(0,0)$.
  - Para $\sqrt[3]{x}$, la intersección es $(0,0)$.
- **Comportamiento**: Varía dependiendo de la forma específica de la función polinómica bajo la raíz.
- **Monotonía**:
  - Para $\sqrt{x}$, la función es creciente.
  - Para $\sqrt[3]{x}$, la función es creciente.
- **Convexidad**:
  - Para $\sqrt{x}$, la función es cóncava hacia abajo en $x > 0$.
  - Para $\sqrt[3]{x}$, la función es cóncava hacia arriba en $x > 0$ y cóncava hacia abajo en $x < 0$.
- **Aplicaciones**: Modelado de fenómenos físicos donde las cantidades están restringidas a valores no negativos 📏⚖️, geometría y análisis de formas curvas 🔄, entre otros.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función irracional
def irrational_function(x):
    return np.sqrt(x)

# Datos de ejemplo
x = np.linspace(0, 5, 400)

# Calcular los valores de y
y = irrational_function(x)

# Graficar la función irracional
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = \sqrt{x}$')
plt.title('Función Irracional')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## 🔍**Funciones Trigonométricas**
Las funciones trigonométricas incluyen seno, coseno, tangente, cotangente, secante y cosecante. Son funciones periódicas que describen relaciones angulares y longitudes en triángulos.

- **Tipos**:
  - Seno: $f(x) = \sin(x)$
  - Coseno: $f(x) = \cos(x)$
  - Tangente: $f(x) = \tan(x)$
  - Cotangente: $f(x) = \cot(x)$
  - Secante: $f(x) = \sec(x)$
  - Cosecante: $f(x) = \csc(x)$
- **Dominio**: Todos los números reales, con exclusiones para tangente, cotangente, secante y cosecante en los puntos donde estas funciones no están definidas.
  - $\sin(x)$ y $\cos(x)$: Todos los números reales.
  - $\tan(x)$ y $\sec(x)$: Todos los números reales excepto $x = \frac{\pi}{2} + k\pi$.
  - $\cot(x)$ y $\csc(x)$: Todos los números reales excepto $x = k\pi$.
- **Recorrido**:
  - Seno y Coseno: $[-1, 1]$.
  - Tangente y Cotangente: Todos los números reales.
  - Secante y Cosecante: $(-\infty, -1] \cup [1, +\infty)$.
- **Periodicidad**: Repetición regular de valores a intervalos específicos.
  - $\sin(x)$ y $\cos(x)$: Período $2\pi$.
  - $\tan(x)$ y $\cot(x)$: Período $\pi$.
  - $\sec(x)$ y $\csc(x)$: Período $2\pi$.
- **Intersecciones con los Ejes**: Dependientes de la función específica y su periodicidad.
  - $\sin(x)$: Intersección en $(k\pi, 0)$.
  - $\cos(x)$: Intersección en $(\frac{\pi}{2} + k\pi, 0)$.
  - $\tan(x)$: Intersección en $(k\pi, 0)$.
- **Comportamiento Asintótico**:
  - $\sin(x)$ y $\cos(x)$ no tienen asíntotas.
  - $\tan(x)$ y $\sec(x)$ tienen asíntotas verticales en $x = \frac{\pi}{2} + k\pi$.
  - $\cot(x)$ y $\csc(x)$ tienen asíntotas verticales en $x = k\pi$.
- **Aplicaciones**: Modelado de fenómenos periódicos como movimiento oscilatorio 🔄, ondas y vibraciones 🌊, análisis de señales 📡, geometría de triángulos 🔺, astronomía 🌌, entre otros.




```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función trigonométrica
def trigonometric_function(x, type='sin'):
    if type == 'sin':
        return np.sin(x)
    elif type == 'cos':
        return np.cos(x)
    elif type == 'tan':
        return np.tan(x)
    elif type == 'cot':
        return 1 / np.tan(x)
    elif type == 'sec':
        return 1 / np.cos(x)
    elif type == 'csc':
        return 1 / np.sin(x)
    else:
        raise ValueError('Tipo de función trigonométrica no válido')

# Datos de ejemplo
x = np.linspace(-2*np.pi, 2*np.pi, 400)

# Calcular los valores de y (por ejemplo, seno)
y = trigonometric_function(x, type='sin')

# Graficar la función trigonométrica (por ejemplo, seno)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = \sin(x)$')
plt.title('Función Trigonométrica (Seno)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

### 🔍**Funciones Polinómicas**
Las funciones polinómicas son sumas de potencias de $x$ con coeficientes constantes y tienen la forma $f(x) = a_n x^n + a_{n-1} x^{n-1} + \ldots + a_1 x + a_0$, donde $a_i$ son coeficientes constantes y $n$ es el grado del polinomio.

- **Dominio**: Todos los números reales.
- **Recorrido**: Determinado por el grado y los coeficientes del polinomio. En general, $(-\infty, +\infty)$.
- **Intersección con el Eje y**: $(0, a_0)$.
- **Raíces**: Los valores de $x$ donde $f(x) = 0$. Pueden ser reales o complejas.
- **Crecimiento**:
  - Polinomios de grado par: Crecen hacia $+\infty$ en ambos extremos o decrecen hacia $-\infty$ en ambos extremos.
  - Polinomios de grado impar: Crecen hacia $+\infty$ en un extremo y decrecen hacia $-\infty$ en el otro.
- **Comportamiento Asintótico**: Determinado por el término de mayor grado. La función se comporta como $a_n x^n$ para valores grandes de $|x|$.
- **Monotonía y Extremos**:
  - Los puntos críticos se encuentran derivando la función y resolviendo $f'(x) = 0$.
  - La naturaleza de estos puntos (máximos, mínimos o puntos de inflexión) se determina con la segunda derivada.
- **Convexidad**:
  - Cóncava hacia arriba (convexa) donde la segunda derivada $f''(x) > 0$.
  - Cóncava hacia abajo (cóncava) donde la segunda derivada $f''(x) < 0$.
- **Aplicaciones**: Ampliamente utilizado en modelado matemático de fenómenos físicos 📊, económicos 💹 y sociales 🌍, resolución de ecuaciones algebraicas 🔍, interpolación y aproximación de funciones 📈, entre otros.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función polinómica
def polynomial_function(x, coeffs):
    return np.polyval(coeffs, x)

# Coeficientes del polinomio de ejemplo: f(x) = 2x^3 - x^2 + 3x - 5
coeffs = [2, -1, 3, -5]

# Datos de ejemplo
x = np.linspace(-2, 2, 400)

# Calcular los valores de y
y = polynomial_function(x, coeffs)

# Graficar la función polinómica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = 2x^3 - x^2 + 3x - 5$')
plt.title('Función Polinómica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

# **Funciones de Utilidad en Economía**💰

En economía, la función de utilidad representa la relación entre las preferencias de un consumidor y las combinaciones de bienes y servicios que maximizan su satisfacción. Esta función es crucial en la teoría del consumidor para explicar cómo los individuos toman decisiones de consumo racionalmente.

## **Características y Propiedades Avanzadas**

## **Definición Matemática**

La función de utilidad $ U(x_1, x_2, \ldots, x_n) $ describe la utilidad total obtenida al consumir cantidades específicas de diferentes bienes $ x_1, x_2, \ldots, x_n $. Puede modelarse con diversas formas funcionales como cuadráticas, logarítmicas, funciones de Cobb-Douglas, entre otras, dependiendo de las preferencias del consumidor.

### **Propiedades Avanzadas:**

- **Convexidad:** La función de utilidad suele ser convexa, lo que significa que las preferencias del consumidor exhiben aversión al riesgo y prefieren una combinación diversificada de bienes.
  
- **Curvas de Indiferencia:** Representan diferentes combinaciones de bienes que proporcionan la misma utilidad al consumidor, reflejando sus preferencias.

### **Teoría de la Elección del Consumidor**💸

Utilizando la función de utilidad, los consumidores maximizan su utilidad sujeta a una restricción presupuestaria, que muestra todas las combinaciones posibles de bienes que el consumidor puede comprar dado su ingreso y los precios de los bienes.

#### **Aplicaciones Avanzadas**

- **Equilibrio del Consumidor:** En la economía de bienestar, la función de utilidad se utiliza para determinar el equilibrio general de la economía, donde la asignación de recursos maximiza la utilidad total de la sociedad.
  
- **Utilidad Esperada:** En decisiones bajo incertidumbre, la función de utilidad se extiende para incluir la utilidad esperada, considerando tanto la utilidad de los resultados como las probabilidades de que ocurran.

## **Funciones de Producción en Economía**🛠️

Las funciones de producción describen cómo se combinan los insumos (como el trabajo y el capital) para producir una cantidad determinada de bienes y servicios. Son esenciales para entender la eficiencia productiva y la asignación de recursos en las empresas y la economía en general.

## **Características y Propiedades Avanzadas**

#### **Definición Matemática:**

La función de producción $ Q = f(L, K) $ describe la cantidad máxima de producción $ Q $ que una empresa puede obtener utilizando ciertas cantidades de insumos de trabajo $ L $ y capital $ K $. Puede tomar diversas formas, como funciones de producción lineales, cuadráticas, Cobb-Douglas, etc., dependiendo de la tecnología y las economías de escala involucradas.

#### **Propiedades y Rendimientos**

- **Rendimientos a Escala:** Determina cómo cambia la producción cuando se aumentan proporcionalmente todos los insumos.
  
- **Rendimientos Marginales:** Indica cómo cambia la producción al aumentar una unidad adicional de un insumo mientras se mantienen constantes los otros.

## **Teoría de la Producción**

Utilizando la función de producción, las empresas maximizan sus beneficios minimizando costos, eligiendo la combinación óptima de insumos para producir una cantidad deseada de producción.

## **Aplicaciones Avanzadas**

- **Economías y Deseconomías de Escala:** La función de producción se utiliza para analizar las economías de escala, donde los costos promedio disminuyen a medida que la producción aumenta, así como las deseconomías de escala donde los costos pueden aumentar debido a la falta de eficiencia.

---

## **Ejemplo Aplicado**

Para ilustrar, consideremos una función de producción Cobb-Douglas típica en economía:

$Q = A \cdot L^{\alpha} \cdot K^{\beta}$

Donde:
- $Q$ es la producción,
- $L$ es la cantidad de trabajo,
- $K$ es la cantidad de capital,
- $\alpha$ y $\beta$ son parámetros que representan las elasticidades de la producción respecto al trabajo y al capital, respectivamente, y $A$ es una constante de escala.

Esta función permite a las empresas determinar cómo ajustar sus niveles de insumos para maximizar la producción dada una restricción de costos.

### **Conclusión**🏦

Las funciones de utilidad y de producción son pilares fundamentales en la teoría económica moderna, proporcionando herramientas poderosas para entender y modelar el comportamiento de consumidores y productores en una economía.

Al comprender sus propiedades avanzadas y aplicaciones prácticas, podemos analizar de manera más rigurosa la asignación de recursos, la eficiencia productiva y el bienestar económico en diferentes contextos.

Dominar estas funciones es esencial para abordar desafíos económicos complejos y tomar decisiones informadas en política económica y gestión empresarial.



```python
import numpy as np
import matplotlib.pyplot as plt

# Función de utilidad cuadrática
def utilidad_cuadratica(x1, x2):
    return 3 * x1**2 + 2 * x2**2

# Definir rangos de variables
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(0, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
Z_utilidad = utilidad_cuadratica(X1, X2)

# Graficar función de utilidad
plt.figure(figsize=(10, 6))
plt.contour(X1, X2, Z_utilidad, levels=20, cmap='viridis')
plt.colorbar(label='Utilidad')
plt.xlabel('Cantidad de Bien 1')
plt.ylabel('Cantidad de Bien 2')
plt.title('Función de Utilidad Cuadrática')
plt.grid(True)
plt.show()

# Función de producción Cobb-Douglas
def produccion_cobb_douglas(L, K):
    alpha = 0.3
    beta = 0.7
    A = 10
    return A * (L**alpha) * (K**beta)

# Definir rangos de insumos
L = np.linspace(0, 100, 100)
K = np.linspace(0, 100, 100)
L, K = np.meshgrid(L, K)
Z_produccion = produccion_cobb_douglas(L, K)

# Graficar función de producción
plt.figure(figsize=(10, 6))
plt.contour(L, K, Z_produccion, levels=20, cmap='plasma')
plt.colorbar(label='Producción')
plt.xlabel('Trabajo (L)')
plt.ylabel('Capital (K)')
plt.title('Función de Producción Cobb-Douglas')
plt.grid(True)
plt.show()
```

**Función de Utilidad Cuadrática:** Muestra curvas de nivel que representan diferentes combinaciones de bienes que proporcionan la misma utilidad al consumidor.

**Función de Producción Cobb-Douglas:** Muestra curvas de nivel que representan diferentes combinaciones de trabajo y capital que producen la misma cantidad de producción.
