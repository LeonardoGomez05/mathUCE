## ***Conceptos B√°sicos***

### **Proposiciones**

Una **proposici√≥n** es una declaraci√≥n que puede ser verdadera o falsa, pero no ambas. Ejemplos de proposiciones:
- "El cielo es azul." (Proposici√≥n verdadera)
- "2 + 2 = 5." (Proposici√≥n falsa)

### **Conectores L√≥gicos**

Los conectores l√≥gicos son operadores que se usan para combinar proposiciones. Los principales conectores son:
- **Negaci√≥n ($¬¨$)**: Invierte el valor de verdad de una proposici√≥n. Si $P$ es verdadera, $¬¨P$ es falsa. Significa ***No***
- **Conjunci√≥n ($‚àß$)**: Es verdadera solo si ambas proposiciones son verdaderas. $P ‚àß Q$ es verdadera si y solo si $P$ y $Q$ son verdaderas. Significa ***y***
- **Disyunci√≥n ($‚à®$)**: Es verdadera si al menos una de las proposiciones es verdadera. $P ‚à® Q$ es verdadera si $P$ o $Q$ es verdadera. Significa ***√≥***.
- **Implicaci√≥n ($‚Üí$)**: \(P ‚Üí Q\) es falsa solo si \(P\) es verdadera y \(Q\) es falsa. De lo contrario, es verdadera. Significa ***Entonces***
- **Doble implicaci√≥n ($‚Üî$)**: $P ‚Üî Q$ es verdadera si $P$ y $Q$ son ambas verdaderas o ambas falsas. Significa ***S√≠ y solo s√≠***

### **Tablas de Verdad** üåü

Las **tablas de verdad** se utilizan para mostrar el valor de verdad de una proposici√≥n compuesta para todas las combinaciones posibles de verdad de sus componentes. Ejemplos de tablas de verdad:

#### ***Negaci√≥n (¬¨)***
| P | ¬¨P |
|---|----|
| V | F  |
| F | V  |

#### ***Conjunci√≥n (‚àß)***
| P | Q | P ‚àß Q |
|---|---|-------|
| V | V |   V   |
| V | F |   F   |
| F | V |   F   |
| F | F |   F   |

#### ***Disyunci√≥n (‚à®)***
| P | Q | P ‚à® Q |
|---|---|-------|
| V | V |   V   |
| V | F |   V   |
| F | V |   V   |
| F | F |   F   |

#### ***Implicaci√≥n (‚Üí)***
| P | Q | P ‚Üí Q |
|---|---|-------|
| V | V |   V   |
| V | F |   F   |
| F | V |   V   |
| F | F |   V   |

#### ***Doble implicaci√≥n (‚Üî)***
| P | Q | P ‚Üî Q |
|---|---|-------|
| V | V |   V   |
| V | F |   F   |
| F | V |   F   |
| F | F |   V   |

## **Paso a Paso para Crear una Tabla de Verdad** üìù

1. **Identificar las Variables Proposicionales:**
   Determina las variables proposicionales que se encuentran en la f√≥rmula l√≥gica. Por ejemplo, en la f√≥rmula $p ‚àß q$, las variables son $p$ y $q$.

2. **Determinar el N√∫mero de Filas:**
   El n√∫mero de filas de la tabla ser√° $2^n$, donde $n$ es el n√∫mero de variables proposicionales. Esto es porque cada variable puede ser verdadera (V) o falsa (F), y necesitas considerar todas las combinaciones posibles.

3. **Listar Todas las Combinaciones de Valores de Verdad:**
   En las primeras columnas de la tabla, lista todas las posibles combinaciones de valores de verdad para las variables proposicionales.

4. **Evaluar la F√≥rmula para Cada Combinaci√≥n:**
   A√±ade columnas adicionales para los subcomponentes de la f√≥rmula y eval√∫a su valor de verdad para cada combinaci√≥n de valores de las variables. Finalmente, eval√∫a la f√≥rmula completa.

## **Tipos de Proposiciones**üìò

### ***Tautolog√≠a*** üü¢

Una tautolog√≠a es una f√≥rmula l√≥gica que es verdadera en todos los posibles valores de verdad de sus componentes. En otras palabras, una tautolog√≠a es una expresi√≥n que no puede ser falsa, sin importar c√≥mo se asignen los valores de verdad a las variables que la componen.

Ejemplos:

$p ‚à® ¬¨p$ (p o no p): Esta es una tautolog√≠a porque, independientemente de si $ùëù$ es verdadero o falso, la expresi√≥n siempre ser√° verdadera.

$(p ‚àß q) ‚Üí (q ‚àß p)$ (si $ùëù$ y $ùëû$ entonces $ùëû$ y $ùëù$): Esta tambi√©n es una tautolog√≠a porque ambas combinaciones ($p$ y $q$ verdaderos o $p$ y $q$ falsos) resultan en una expresi√≥n verdadera.

Las tautolog√≠as son fundamentales en l√≥gica y matem√°ticas porque representan verdades universales. Son √∫tiles para probar teoremas y propiedades en diversas √°reas del conocimiento.

### ***Contradicci√≥n*** üî¥

Una contradicci√≥n es una f√≥rmula l√≥gica que es falsa en todos los posibles valores de verdad de sus componentes. Es una expresi√≥n que no puede ser verdadera bajo ninguna asignaci√≥n de valores de verdad.

Ejemplos:

$p ‚àß ¬¨p$ (p y no p): Esta es una contradicci√≥n porque es imposible que
$ùëù$ sea verdadero y falso al mismo tiempo.

(ùëù ‚Üí ùëû) ‚àß (ùëù ‚àß ¬¨ùëû) (si ùëù entonces ùëû y ùëù y no ùëû): Esta expresi√≥n es siempre falsa porque no puede ser que ùëù implique ùëû y al mismo tiempo ùëû sea falso cuando ùëù es verdadero.

Las contradicciones son importantes en l√≥gica porque representan situaciones imposibles. En la demostraci√≥n de teoremas, llegar a una contradicci√≥n puede indicar que una suposici√≥n inicial era incorrecta (principio de reducci√≥n al absurdo).

### ***Contingencia*** üü°

Una contingencia es una f√≥rmula l√≥gica que puede ser verdadera o falsa dependiendo de los valores de verdad de sus componentes. No es una tautolog√≠a ni una contradicci√≥n, es decir, su valor de verdad depende de la asignaci√≥n espec√≠fica de valores a las variables.

Ejemplos:

$ùëù ‚àß ùëû$ (p y q): Esta es una contingencia porque puede ser verdadera si tanto ùëù como ùëû son verdaderos, pero ser√° falsa si al menos una de las dos es falsa.

$ùëù ‚Üí ùëû$ (si ùëù entonces ùëû): Esta expresi√≥n puede ser verdadera o falsa dependiendo de los valores de $ùëù$ y $ùëû$. Ser√° falsa si $ùëù$ es verdadero y $ùëû$ es falso; en todos los dem√°s casos ser√° verdadera.

Las contingencias son fundamentales para el razonamiento l√≥gico y la toma de decisiones. Reflejan situaciones en las que el resultado no es absoluto y depende de circunstancias espec√≠ficas.

### **Resumen**

***Tautolog√≠a:*** Siempre verdadera, independientemente de los valores de verdad de sus componentes. üü¢

***Contradicci√≥n:*** Siempre falsa, independientemente de los valores de verdad de sus componentes. üî¥

***Contingencia:*** Puede ser verdadera o falsa, dependiendo de los valores de verdad de sus componentes. üü°

Estas tres categor√≠as son esenciales en la l√≥gica proposicional y forman la base para entender c√≥mo se eval√∫an y manipulan las proposiciones l√≥gicas. üß©

## **Argumentos L√≥gicos**üåê

Un **argumento** es un conjunto de proposiciones, donde una de ellas es la **conclusi√≥n** y las dem√°s son las **premisas**. Un argumento es v√°lido si la conclusi√≥n se sigue necesariamente de las premisas.

### **Formas Argumentativas Comunes**

#### ***Modus Ponens*** üü©

Si $P ‚Üí Q$ y $P$, entonces $Q$.
- Ejemplo:
  - Premisa 1: Si llueve, la calle estar√° mojada.
  - Premisa 2: Llueve.
  - Conclusi√≥n: La calle est√° mojada.

#### ***Modus Tollens*** üü•

Si $P ‚Üí Q$ y $¬¨Q$, entonces $¬¨P$.
- Ejemplo:
  - Premisa 1: Si llueve, la calle estar√° mojada.
  - Premisa 2: La calle no est√° mojada.
  - Conclusi√≥n: No llueve.

#### ***Silogismo Disyuntivo*** üî∂

Si $P ‚à® Q$ y $¬¨P$, entonces $Q$.
- Ejemplo:
  - Premisa 1: O llueve o hace sol.
  - Premisa 2: No llueve.
  - Conclusi√≥n: Hace sol.

#### ***Silogismo Hipot√©tico*** üî∑

Si $P ‚Üí Q$ y $Q ‚Üí R$, entonces $P ‚Üí R$.
- Ejemplo:
  - Premisa 1: Si estudio, aprobar√© el examen.
  - Premisa 2: Si apruebo el examen, pasar√© el curso.
  - Conclusi√≥n: Si estudio, pasar√© el curso.

## **L√≥gica Cuantificacional** üåç

La **l√≥gica cuantificacional** extiende la l√≥gica proposicional para incluir cuantificadores y predicados.

### ***Cuantificadores*** üî¢

- **Cuantificador Universal (‚àÄ)**: "Para todo" o "Para cada". $‚àÄx P(x)$ significa que $P(x)$ es verdadera para todo $x$.
- **Cuantificador Existencial (‚àÉ)**: "Existe" o "Al menos uno". $‚àÉx P(x)$ significa que hay al menos un $x$ para el cual $P(x)$ es verdadera.

### Ejemplos üß©

1. $‚àÄx (P(x) ‚Üí Q(x))$: Para todo $x$, si $P(x)$ es verdadera, entonces $Q(x)$ es verdadera.
2. $‚àÉx (P(x) ‚àß Q(x))$: Existe al menos un $x$ tal que $P(x)$ y $Q(x)$ son verdaderas.

## **L√≥gica de Predicados** üî¨

### ***Predicados*** üßÆ

Un **predicado** es una funci√≥n que toma uno o m√°s argumentos y devuelve un valor de verdad. Por ejemplo, $P(x)$ podr√≠a significar "$x$ es un n√∫mero par".

### Ejemplos de L√≥gica de Predicados

1. $P(x)$ = "x es mayor que 3"
   - $P(4)$ es verdadero.
   - $P(2)$ es falso.

2. $Q(x, y)$ = "x es mayor que y"
   -  $Q(5, 3)$ es verdadero.
   -  $Q(2, 4)$ es falso.

### **Reglas de Inferencia**

Las reglas de inferencia son las reglas que se utilizan para derivar conclusiones v√°lidas a partir de premisas. Algunas de las reglas m√°s comunes son:

- **Regla de la Conjunci√≥n**: De $P$ y $Q$, se puede inferir $P ‚àß Q$.
- **Regla de la Disyunci√≥n**: De $P$, se puede inferir $P ‚à® Q$.
- **Regla de la Simplificaci√≥n**: De $P ‚àß Q$, se puede inferir $P$.
- **Regla de la Resoluci√≥n**: De $P ‚à® Q$ y $¬¨P ‚à® R$, se puede inferir $Q ‚à® R$.

## **M√©todos de Prueba**

### ***Prueba Directa***

En una **prueba directa**, se asume que las premisas son verdaderas y se utiliza una serie de pasos l√≥gicos para demostrar que la conclusi√≥n es verdadera.

### ***Prueba por Contradicci√≥n***

En una **prueba por contradicci√≥n**, se asume que la conclusi√≥n es falsa y se demuestra que esta suposici√≥n lleva a una contradicci√≥n con las premisas, lo que implica que la conclusi√≥n debe ser verdadera.

### ***Prueba por Inducci√≥n***

La **prueba por inducci√≥n** es una t√©cnica usada principalmente en matem√°ticas para demostrar que una proposici√≥n es verdadera para todos los n√∫meros naturales. Consiste en dos pasos:
1. **Base**: Probar que la proposici√≥n es verdadera para el primer n√∫mero natural.
2. **Paso Inductivo**: Probar que si la proposici√≥n es verdadera para un n√∫mero natural \(k\), entonces tambi√©n es verdadera para \(k + 1\).

```python
# Operaciones l√≥gicas b√°sicas

# Definici√≥n de valores booleanos
a = True
b = False

# Operaci√≥n l√≥gica AND: retorna True si ambos operandos son True
and_result = a and b
print(f"AND: {a} and {b} = {and_result}")

# Operaci√≥n l√≥gica OR: retorna True si al menos uno de los operandos es True
or_result = a or b
print(f"OR: {a} or {b} = {or_result}")

# Operaci√≥n l√≥gica NOT: retorna True si el operando es False y viceversa
not_result = not a
print(f"NOT: not {a} = {not_result}")

# Operaci√≥n l√≥gica XOR: retorna True si los operandos son diferentes
xor_result = a ^ b
print(f"XOR: {a} ^ {b} = {xor_result}")
```

```python
# Verificaci√≥n de tautolog√≠as

# Funci√≥n para verificar si una expresi√≥n es una tautolog√≠a
def es_tautologia(expr):
    # Lista de posibles valores booleanos
    valores = [False, True]
    for p in valores:
        for q in valores:
            if not eval(expr):
                return False
    return True

# Expresi√≥n l√≥gica (P or not P) que es una tautolog√≠a
expresion = "(p or not p)"
print(f"La expresi√≥n {expresion} es una tautolog√≠a: {es_tautologia(expresion)}")

# Expresi√≥n l√≥gica (P and not P) que no es una tautolog√≠a
expresion = "(p and not p)"
print(f"La expresi√≥n {expresion} es una tautolog√≠a: {es_tautologia(expresion)}")
```

```python
# Demostraci√≥n de contradicciones

# Funci√≥n para verificar si una expresi√≥n es una contradicci√≥n
def es_contradiccion(expr):
    # Lista de posibles valores booleanos
    valores = [False, True]
    for p in valores:
        for q in valores:
            if eval(expr):
                return False
    return True

# Expresi√≥n l√≥gica (P and not P) que es una contradicci√≥n
expresion = "(p and not p)"
print(f"La expresi√≥n {expresion} es una contradicci√≥n: {es_contradiccion(expresion)}")

# Expresi√≥n l√≥gica (P or not P) que no es una contradicci√≥n
expresion = "(p or not p)"
print(f"La expresi√≥n {expresion} es una contradicci√≥n: {es_contradiccion(expresion)}")
```

```python
# Cuantificadores usando listas y funciones

# Funci√≥n para verificar si todos los elementos de una lista son positivos
def todos_positivos(lista):
    return all(x > 0 for x in lista)

# Lista de n√∫meros
numeros = [1, 2, 3, 4, 5]

# Verificar si todos los elementos son positivos
resultado = todos_positivos(numeros)
print(f"Todos los elementos son positivos: {resultado}")

# Funci√≥n para verificar si existe al menos un elemento negativo en la lista
def existe_negativo(lista):
    return any(x < 0 for x in lista)

# Lista de n√∫meros con un elemento negativo
numeros = [1, 2, 3, -4, 5]

# Verificar si existe al menos un elemento negativo
resultado = existe_negativo(numeros)
print(f"Existe al menos un elemento negativo: {resultado}")
```

# **Tipos de Conjuntos**

1. **Conjunto Vac√≠o (o Conjunto Nulo):**
   - Definici√≥n: Conjunto que no contiene elementos.
   - S√≠mbolo: $ \emptyset $ o $ \{\} $.
   - Ejemplo: $ \emptyset = \{\} $.

2. **Conjunto Unitario:**
   - Definici√≥n: Conjunto que contiene exactamente un elemento.
   - S√≠mbolo: $ \{a\} $, donde $ a $ es un elemento.
   - Ejemplo: $ \{1\} $, $ \{a\} $.

3. **Conjunto Finito:**
   - Definici√≥n: Conjunto que contiene un n√∫mero finito de elementos.
   - Ejemplo: $ \{1, 2, 3, 4, 5\} $.

4. **Conjunto Infinito:**
   - Definici√≥n: Conjunto que contiene un n√∫mero infinito de elementos.
   - Ejemplo: $ \mathbb{N} = \{1, 2, 3, \dots\} $ (n√∫meros naturales).

5. **Conjunto Unitario de Conjuntos:**
   - Definici√≥n: Conjunto que contiene un solo conjunto como elemento.
   - Ejemplo: $ \{\{1, 2\}\} $.

6. **Conjunto de Partes (o Conjunto Potencia):**
   - Definici√≥n: Conjunto que contiene todos los subconjuntos de otro conjunto dado.
   - Ejemplo: Si $ A = \{1, 2\} $, entonces $ \mathcal{P}(A) = \{\emptyset, \{1\}, \{2\}, \{1, 2\}\} $.

7. **Conjunto Disjunto:**
   - Definici√≥n: Conjunto cuyos elementos no tienen elementos en com√∫n con otro conjunto dado.
   - Ejemplo: Si $ A = \{1, 2\} $ y $ B = \{3, 4\} $, entonces $ A $ y $ B $ son conjuntos disjuntos.

8. **Conjunto de N√∫meros (o Conjunto Num√©rico):**
   - Definici√≥n: Conjunto que contiene n√∫meros de un tipo espec√≠fico, como n√∫meros naturales, enteros, racionales, etc.
   - Ejemplo: $ \mathbb{N} = \{1, 2, 3, \dots\} $ (n√∫meros naturales).

9. **Conjunto Singleton:**
   - Definici√≥n: Conjunto que contiene exactamente un elemento.
   - Ejemplo: $ \{5\} $.

10. **Conjunto Infinito Enumerado:**
    - Definici√≥n: Conjunto que contiene una cantidad infinita de elementos que pueden ser listados en orden.
    - Ejemplo: $ \mathbb{Z} = \{\dots, -3, -2, -1, 0, 1, 2, 3, \dots\} $ (n√∫meros enteros).

## **Operaciones con Conjuntos**
### ***Uni√≥n***
La uni√≥n de dos conjuntos $A$ y $B$ es el conjunto de elementos que est√°n en $A$, en $B$, o en ambos. Se denota por $A ‚à™ B$.

$ A ‚à™ B = \{ x \mid x ‚àà A \text{ o } x ‚àà B \}$

   - Ejemplo:
     - Si $A$ = {$ 1, 2, 3 $} y $B$ = {$3, 4, 5$},
     - Entonces $A \cup B =$ {$1, 2, 3, 4, 5$}.

### ***Intersecci√≥n***
La intersecci√≥n de dos conjuntos $A$ y $B$ es el conjunto de elementos que est√°n tanto en $A$ como en $B$. Se denota por $A ‚à© B$.

$A ‚à© B = \{ x \mid x ‚àà A \text{ y } x ‚àà B \}$

   - Ejemplo:
     - Si $A$ = {$1, 2, 3$} y $B$ = {$3, 4, 5$},
     - Entonces $A \cap B = \{3\}$.

### ***Diferencia***
La diferencia de dos conjuntos $A$ y $B$ es el conjunto de elementos que est√°n en A pero no en B. Se denota por $A - B$.

$A - B = \{ x \mid x ‚àà A \text{ y } x ‚àâ B \}$

   - Ejemplo:
     - Si $A$ = {$1, 2, 3$} y $B$ = {$3, 4, 5$},
     - Entonces $A \setminus B = \{1, 2\}$.

### ***Complemento***
El complemento de un conjunto $A$, relativo a un conjunto universal $U$, es el conjunto de elementos que no est√°n en $A$. Se denota por $A'$ o por $A^c$.


$A' = \{ x \mid x ‚àâ A \}$

   - Ejemplo (suponiendo un conjunto universal $U = \{1, 2, 3, 4, 5\}$):
     - Si $A$ = {$1, 2, 3$},
     - Entonces $A^c = U \setminus A =$ {$4, 5$}.

# **Relaci√≥n de Contenci√≥n**

La relaci√≥n de contenci√≥n entre dos conjuntos $A$ y $B$ se define como $A \subseteq B$ si y solo si cada elemento de $A$ es tambi√©n un elemento de $B$. Esto se puede expresar formalmente como:

$$ \forall x \, (x \in A \Rightarrow x \in B) $$

Esto significa que para todo elemento $x$ en $A$, tambi√©n se cumple que $x$ est√° en $B$. En caso de que $A$ sea igual a $B$, se utiliza el s√≠mbolo $A = B$, lo que indica que ambos conjuntos contienen exactamente los mismos elementos.

## **Propiedades Importantes**

1. **Reflexividad:** Todo conjunto $A$ es un subconjunto de s√≠ mismo, es decir, $A \subseteq A$.
   
2. **Antisimetr√≠a:** Si $A \subseteq B$ y $B \subseteq A$, entonces $A = B$. Esto implica que dos conjuntos son iguales si y solo si son subconjuntos uno del otro.

3. **Transitividad:** Si $A \subseteq B$ y $B \subseteq C$, entonces $A \subseteq C$. Es decir, si $A$ est√° contenido en $B$ y $B$ est√° contenido en $C$, entonces $A$ tambi√©n est√° contenido en $C$.

## **Ejemplos Pr√°cticos**

1. **Conjuntos Num√©ricos:**
   - El conjunto de los n√∫meros naturales $\mathbb{N} = \{1, 2, 3, \ldots\}$ est√° contenido en el conjunto de los n√∫meros enteros
   $\mathbb{Z} = \{\ldots, -2, -1, 0, 1, 2, \ldots\}$.
   - El conjunto de los n√∫meros enteros $\mathbb{Z}$ est√° contenido en el conjunto de los n√∫meros racionales $\mathbb{Q}$.
   - El conjunto de los n√∫meros racionales $\mathbb{Q}$ est√° contenido en el conjunto de los n√∫meros reales $\mathbb{R}$.
   -En resumen, $\mathbb{N} \subseteq \mathbb{Z} \subseteq \mathbb{Q} \subseteq \mathbb{R}$

2. **Subconjuntos en Conjuntos Finitos:**
   - Si tenemos dos conjuntos finitos $A$ y $B$, podemos verificar la relaci√≥n de contenci√≥n contando los elementos de $A$ y $B$. Por ejemplo, si $A = \{1, 2\}$ y $B = \{1, 2, 3\}$, entonces $A \subseteq B$.

3. **Conjuntos Vac√≠os:**
   - El conjunto vac√≠o $\emptyset$ es un subconjunto de cualquier conjunto, es decir, $\emptyset \subseteq A$ para cualquier conjunto $A$.





### **Propiedades de los Conjuntos**

1. **Conjunto Vac√≠o:**
   - Propiedad: Todo conjunto es subconjunto del conjunto vac√≠o.
   - $ \forall A, A \subseteq \emptyset $.

2. **Uni√≥n:**
   - Propiedad Conmutativa: $ A \cup B = B \cup A $.
   - Propiedad Asociativa: $ A \cup (B \cup C) = (A \cup B) \cup C $.
   - Propiedad de Identidad: $ A \cup \emptyset = A $.
   - Propiedad de Absorci√≥n: $ A \cup A = A $.

3. **Intersecci√≥n:**
   - Propiedad Conmutativa: $ A \cap B = B \cap A $.
   - Propiedad Asociativa: $ A \cap (B \cap C) = (A \cap B) \cap C $.
   - Propiedad de Identidad: $ A \cap U = A $ (donde $ U $ es el conjunto universal).
   - Propiedad de Absorci√≥n: $ A \cap (A \cup B) = A $.

4. **Diferencia:**
   - Propiedad de la Diferencia con el Conjunto Vac√≠o: $ A \setminus \emptyset = A $.
   - Propiedad de Idempotencia: $ A \setminus A = \emptyset $.

5. **Complemento:**
   - Propiedad de Complemento con el Conjunto Universal: $ A \cup A' = U $.
   - Propiedad de Complemento con el Conjunto Vac√≠o: $ A \cap A' = \emptyset $.

6. **Propiedades Distributivas:**
   - Distributiva de la Intersecci√≥n sobre la Uni√≥n: $ A \cap (B \cup C) = (A \cap B) \cup (A \cap C) $.
   - Distributiva de la Uni√≥n sobre la Intersecci√≥n: $ A \cup (B \cap C) = (A \cup B) \cap (A \cup C) $.

7. **Leyes de Morgan:**
   - Ley de Morgan para la Negaci√≥n de la Uni√≥n: $ (A \cup B)' = A' \cap B' $.
   - Ley de Morgan para la Negaci√≥n de la Intersecci√≥n: $ (A \cap B)' = A' \cup B' $.

8. **Propiedades de Subconjuntos:**
   - Propiedad Reflexiva: $ A \subseteq A $.
   - Propiedad Transitiva: Si $ A \subseteq B $ y $ B \subseteq C $, entonces $ A \subseteq C $.
   - Propiedad de la Igualdad de Conjuntos: Dos conjuntos son iguales si y solo si tienen los mismos elementos.

9. **Propiedades de Cardinalidad:**
   - Propiedad de la Cardinalidad de la Uni√≥n: $ |A \cup B| = |A| + |B| - |A \cap B| $.
   - Propiedad de la Cardinalidad de la Intersecci√≥n: $ |A \cap B| = |A| + |B| - |A \cup B| $.

# **Producto Cartesiano**

El producto cartesiano es una operaci√≥n fundamental en la teor√≠a de conjuntos que combina elementos de dos conjuntos diferentes para formar pares ordenados. Este concepto es crucial en matem√°ticas, especialmente en √°lgebra, geometr√≠a y teor√≠a de n√∫meros.

## **Definici√≥n**

Dado dos conjuntos $ A $ y $ B $, el producto cartesiano $ A \times B $ es el conjunto de todos los pares ordenados $ (a, b) $, donde $ a \in A $ y $ b \in B $.

## **Notaci√≥n**

- El producto cartesiano se denota como $ A \times B $ o $ (A, B) $.
- Si $ A = \{a_1, a_2, \ldots, a_m\} $ y $ B = \{b_1, b_2, \ldots, b_n\} $, entonces
  $$ A \times B = \{ (a_1, b_1), (a_1, b_2), \ldots, (a_m, b_n) \}. $$

## Ejemplos

1. **Ejemplo Simple:**
   - Si $ A = \{1, 2\} $ y $ B = \{a, b, c\} $, entonces

     $ A \times B = \{(1, a), (1, b), (1, c), (2, a), (2, b), (2, c)\} $.

2. **Ejemplo Geom√©trico:**
   - En geometr√≠a, el producto cartesiano $ \mathbb{R}^2 = \mathbb{R} \times \mathbb{R} $ representa el plano euclidiano.
